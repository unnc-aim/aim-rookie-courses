```python
# æ¡ä»¶è¯­å¥è¯¦è§£

print("=== åŸºæœ¬æ¡ä»¶è¯­å¥ ===")

# ç®€å•ifè¯­å¥
robot_hp = 150

if robot_hp > 200:
    status = "çŠ¶æ€ä¼˜ç§€"
    action = "ç»§ç»­æˆ˜æ–—"
elif robot_hp > 100:
    status = "çŠ¶æ€è‰¯å¥½" 
    action = "å°å¿ƒåº”æˆ˜"
elif robot_hp > 50:
    status = "çŠ¶æ€ä¸€èˆ¬"
    action = "å¯»æ‰¾æ©æŠ¤"
else:
    status = "çŠ¶æ€å±æ€¥"
    action = "ç«‹å³æ’¤é€€"

print(f"æœºå™¨äººè¡€é‡: {robot_hp}")
print(f"çŠ¶æ€è¯„ä¼°: {status}")
print(f"å»ºè®®è¡ŒåŠ¨: {action}")

# å¤æ‚æ¡ä»¶åˆ¤æ–­
print(f"\\n=== å¤æ‚æ¡ä»¶åˆ¤æ–­ ===")

def analyze_robot_status(hp, armor, ammo, energy):
    """åˆ†ææœºå™¨äººç»¼åˆçŠ¶æ€"""
    
    print(f"åˆ†æå‚æ•°: HP={hp}, æŠ¤ç”²={armor}, å¼¹è¯={ammo}, èƒ½é‡={energy}")
    
    # å¤šæ¡ä»¶ç»„åˆåˆ¤æ–­
    if hp <= 0:
        return "æœºå™¨äººå·²æŸå", "éœ€è¦ç»´ä¿®"
    
    if hp > 150 and armor > 50 and ammo > 100 and energy > 70:
        return "æˆ˜æ–—çŠ¶æ€æä½³", "ä¸»åŠ¨å‡ºå‡»"
    elif hp > 100 and (armor > 30 or ammo > 50) and energy > 50:
        return "æˆ˜æ–—çŠ¶æ€è‰¯å¥½", "è°¨æ…æ¨è¿›"
    elif hp > 50 and energy > 30:
        if ammo <= 20:
            return "å¼¹è¯ä¸è¶³", "è¡¥å……å¼¹è¯"
        elif armor <= 20:
            return "æŠ¤ç”²å—æŸ", "ç»´ä¿®æŠ¤ç”²"
        else:
            return "çŠ¶æ€ä¸€èˆ¬", "ä¿å®ˆä½œæˆ˜"
    else:
        return "çŠ¶æ€å±æ€¥", "ç«‹å³æ’¤é€€"

# æµ‹è¯•ä¸åŒçŠ¶æ€
test_cases = [
    (200, 80, 150, 90),  # æœ€ä½³çŠ¶æ€
    (120, 40, 60, 65),   # è‰¯å¥½çŠ¶æ€
    (80, 15, 30, 45),    # ä¸€èˆ¬çŠ¶æ€
    (30, 5, 10, 20),     # å±æ€¥çŠ¶æ€
    (0, 0, 0, 0)         # æŸåçŠ¶æ€
]

for i, (hp, armor, ammo, energy) in enumerate(test_cases, 1):
    status, action = analyze_robot_status(hp, armor, ammo, energy)
    print(f"æµ‹è¯•{i}: {status} -> {action}")
    print()

# ä¸‰å…ƒè¿ç®—ç¬¦ï¼ˆæ¡ä»¶è¡¨è¾¾å¼ï¼‰
print(f"=== ä¸‰å…ƒè¿ç®—ç¬¦ ===")
robot_energy = 75
battery_status = "å……è¶³" if robot_energy > 50 else "ä¸è¶³"
print(f"èƒ½é‡: {robot_energy}%, ç”µæ± çŠ¶æ€: {battery_status}")

# é“¾å¼æ¯”è¾ƒ
temperature = 25
comfort_level = (
    "è¿‡çƒ­" if temperature > 30 else
    "èˆ’é€‚" if 20 <= temperature <= 30 else
    "è¿‡å†·"
)
print(f"æ¸©åº¦: {temperature}Â°C, èˆ’é€‚åº¦: {comfort_level}")

# matchè¯­å¥ï¼ˆPython 3.10+çš„æ–°ç‰¹æ€§ï¼‰
print(f"\\n=== matchè¯­å¥ç¤ºä¾‹ ===")

def handle_robot_command(command):
    """å¤„ç†æœºå™¨äººæŒ‡ä»¤"""
    
    # æ³¨æ„ï¼šmatchè¯­å¥éœ€è¦Python 3.10+
    # è¿™é‡Œç”¨if-elifæ¨¡æ‹Ÿmatchçš„æ•ˆæœ
    
    if command == "attack":
        return "æ‰§è¡Œæ”»å‡»æŒ‡ä»¤"
    elif command == "defend":
        return "è¿›å…¥é˜²å¾¡æ¨¡å¼"
    elif command == "retreat":
        return "å¼€å§‹æ’¤é€€"
    elif command in ["move_forward", "move_backward", "move_left", "move_right"]:
        return f"æ‰§è¡Œç§»åŠ¨æŒ‡ä»¤: {command}"
    elif command.startswith("target_"):
        target = command.split("_")[1]
        return f"é”å®šç›®æ ‡: {target}"
    else:
        return "æœªçŸ¥æŒ‡ä»¤ï¼Œè¯·é‡æ–°è¾“å…¥"

# æµ‹è¯•æŒ‡ä»¤å¤„ç†
commands = ["attack", "defend", "move_forward", "target_enemy1", "invalid_cmd"]
for cmd in commands:
    result = handle_robot_command(cmd)
    print(f"æŒ‡ä»¤ '{cmd}': {result}")

# æ¡ä»¶è¯­å¥çš„æœ€ä½³å®è·µ
print(f"\\n=== æ¡ä»¶è¯­å¥æœ€ä½³å®è·µ ===")

def evaluate_team_performance(scores) -> dict:
    """è¯„ä¼°é˜Ÿä¼è¡¨ç°"""
    
    if not scores:  # æ£€æŸ¥ç©ºåˆ—è¡¨
        return {
            "performance": "",
            "average": 0.0,
            "range": 0,
            "consistency": "æœªçŸ¥"
        }
    
    avg_score = sum(scores) / len(scores)
    max_score = max(scores)
    min_score = min(scores)
    
    # ä½¿ç”¨å¸¸é‡æé«˜ä»£ç å¯è¯»æ€§
    EXCELLENT_THRESHOLD = 90
    GOOD_THRESHOLD = 75
    AVERAGE_THRESHOLD = 60
    
    if avg_score >= EXCELLENT_THRESHOLD:
        performance = "ä¼˜ç§€"
    elif avg_score >= GOOD_THRESHOLD:
        performance = "è‰¯å¥½"
    elif avg_score >= AVERAGE_THRESHOLD:
        performance = "ä¸€èˆ¬"
    else:
        performance = "éœ€è¦æ”¹è¿›"
    
    # é¢å¤–åˆ†æ
    score_range = max_score - min_score
    consistency = "ç¨³å®š" if score_range <= 20 else "ä¸ç¨³å®š"
    
    return {
        "performance": performance,
        "average": round(avg_score, 1),
        "range": score_range,
        "consistency": consistency
    }

# æµ‹è¯•é˜Ÿä¼è¡¨ç°è¯„ä¼°
team_scores = [88, 92, 85, 90, 87]
evaluation = evaluate_team_performance(team_scores)

print(f"é˜Ÿä¼åˆ†æ•°: {team_scores}")
print(f"è¡¨ç°è¯„ä¼°: {evaluation['performance']}")
print(f"å¹³å‡åˆ†: {evaluation['average']}")
print(f"åˆ†æ•°æ³¢åŠ¨: {evaluation['range']} (çŠ¶æ€: {evaluation['consistency']})")

print("\\nâœ… æ¡ä»¶è¯­å¥å­¦ä¹ å®Œæˆï¼")
```

```python
# Lesson-2 ç»¼åˆå®è·µé¡¹ç›®ï¼šå­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ

print("ğŸ“ å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ")
print("=" * 50)

# ä½¿ç”¨å…¨å±€å˜é‡å­˜å‚¨æ•°æ®
students = {}  # å­˜å‚¨å­¦ç”Ÿä¿¡æ¯
courses = []   # å­˜å‚¨è¯¾ç¨‹åˆ—è¡¨

def add_course(course_name):
    """æ·»åŠ è¯¾ç¨‹"""
    global courses
    if course_name not in courses:
        courses.append(course_name)
        print(f"âœ… å·²æ·»åŠ è¯¾ç¨‹: {course_name}")
    else:
        print(f"âŒ è¯¾ç¨‹ {course_name} å·²å­˜åœ¨")

def add_student(student_id, name):
    """æ·»åŠ å­¦ç”Ÿ"""
    global students
    if student_id not in students:
        students[student_id] = {
            "name": name,
            "grades": {}
        }
        print(f"âœ… å·²æ·»åŠ å­¦ç”Ÿ: {name} (ID: {student_id})")
    else:
        print(f"âŒ å­¦ç”ŸID {student_id} å·²å­˜åœ¨")

def add_grade(student_id, course, score):
    """æ·»åŠ æˆç»©"""
    global students, courses
    if student_id in students:
        if course in courses:
            students[student_id]["grades"][course] = score
            print(f"âœ… å·²å½•å…¥ {students[student_id]['name']} çš„ {course} æˆç»©: {score}")
        else:
            print(f"âŒ è¯¾ç¨‹ {course} ä¸å­˜åœ¨ï¼Œè¯·å…ˆæ·»åŠ è¯¾ç¨‹")
    else:
        print(f"âŒ å­¦ç”ŸID {student_id} ä¸å­˜åœ¨")

def get_student_average(student_id):
    """è®¡ç®—å­¦ç”Ÿå¹³å‡åˆ†"""
    global students
    if student_id in students:
        grades = list(students[student_id]["grades"].values())
        if grades:
            return sum(grades) / len(grades)
        else:
            return 0
    return None

def get_course_statistics(course):
    """è·å–è¯¾ç¨‹ç»Ÿè®¡ä¿¡æ¯"""
    global students
    scores = []
    for student in students.values():
        if course in student["grades"]:
            scores.append(student["grades"][course])
    
    if scores:
        return {
            "count": len(scores),
            "average": sum(scores) / len(scores),
            "max": max(scores),
            "min": min(scores),
            "pass_rate": len([s for s in scores if s >= 60]) / len(scores) * 100
        }
    return None

def display_all_students():
    """æ˜¾ç¤ºæ‰€æœ‰å­¦ç”Ÿä¿¡æ¯"""
    global students
    print(f"\nğŸ“Š å­¦ç”Ÿä¿¡æ¯æ€»è§ˆ:")
    print("-" * 60)
    
    for student_id, info in students.items():
        name = info["name"]
        grades = info["grades"]
        avg = get_student_average(student_id)
        
        print(f"å­¦å·: {student_id} | å§“å: {name} | å¹³å‡åˆ†: {avg:.1f}")
        
        if grades:
            for course, score in grades.items():
                status = "âœ… åŠæ ¼" if score >= 60 else "âŒ ä¸åŠæ ¼"
                print(f"  â””â”€ {course}: {score}åˆ† {status}")
        else:
            print("  â””â”€ æš‚æ— æˆç»©è®°å½•")
        print()

def display_course_statistics():
    """æ˜¾ç¤ºè¯¾ç¨‹ç»Ÿè®¡"""
    global courses
    print(f"\nğŸ“ˆ è¯¾ç¨‹ç»Ÿè®¡ä¿¡æ¯:")
    print("-" * 60)
    
    for course in courses:
        stats = get_course_statistics(course)
        if stats:
            print(f"è¯¾ç¨‹: {course}")
            print(f"  å‚ä¸äººæ•°: {stats['count']}")
            print(f"  å¹³å‡åˆ†: {stats['average']:.1f}")
            print(f"  æœ€é«˜åˆ†: {stats['max']}")
            print(f"  æœ€ä½åˆ†: {stats['min']}")
            print(f"  åŠæ ¼ç‡: {stats['pass_rate']:.1f}%")
            print()
        else:
            print(f"è¯¾ç¨‹: {course} (æš‚æ— æˆç»©æ•°æ®)\n")

def reset_system():
    """é‡ç½®ç³»ç»Ÿæ•°æ®"""
    global students, courses
    students = {}
    courses = []
    print("ğŸ”„ ç³»ç»Ÿæ•°æ®å·²é‡ç½®")

def demo_grade_system():
    """æ¼”ç¤ºæˆç»©ç®¡ç†ç³»ç»Ÿ"""
    
    # é‡ç½®ç³»ç»Ÿ
    reset_system()
    
    # æ·»åŠ è¯¾ç¨‹
    print("1ï¸âƒ£ æ·»åŠ è¯¾ç¨‹")
    course_list = ["Pythonç¼–ç¨‹", "æ•°æ®ç»“æ„", "ç®—æ³•è®¾è®¡", "æœºå™¨å­¦ä¹ "]
    for course in course_list:
        add_course(course)
    
    print(f"\n2ï¸âƒ£ æ·»åŠ å­¦ç”Ÿ")
    # æ·»åŠ å­¦ç”Ÿ
    student_list = [
        ("2024001", "å¼ ä¸‰"),
        ("2024002", "æå››"), 
        ("2024003", "ç‹äº”"),
        ("2024004", "èµµå…­"),
        ("2024005", "å­™ä¸ƒ")
    ]
    
    for student_id, name in student_list:
        add_student(student_id, name)
    
    print(f"\n3ï¸âƒ£ å½•å…¥æˆç»©")
    # å½•å…¥æˆç»©
    grades_data = [
        ("2024001", "Pythonç¼–ç¨‹", 95),
        ("2024001", "æ•°æ®ç»“æ„", 88),
        ("2024001", "ç®—æ³•è®¾è®¡", 92),
        ("2024002", "Pythonç¼–ç¨‹", 78),
        ("2024002", "æ•°æ®ç»“æ„", 85),
        ("2024002", "æœºå™¨å­¦ä¹ ", 90),
        ("2024003", "Pythonç¼–ç¨‹", 56),
        ("2024003", "æ•°æ®ç»“æ„", 72),
        ("2024003", "ç®—æ³•è®¾è®¡", 68),
        ("2024004", "Pythonç¼–ç¨‹", 91),
        ("2024004", "æœºå™¨å­¦ä¹ ", 87),
        ("2024005", "Pythonç¼–ç¨‹", 83),
        ("2024005", "æ•°æ®ç»“æ„", 79),
        ("2024005", "ç®—æ³•è®¾è®¡", 85),
        ("2024005", "æœºå™¨å­¦ä¹ ", 92)
    ]
    
    for student_id, course, score in grades_data:
        add_grade(student_id, course, score)
    
    # æ˜¾ç¤ºç»“æœ
    display_all_students()
    display_course_statistics()

# è¿è¡Œæ¼”ç¤º
demo_grade_system()

print("\n" + "=" * 50)
print("ğŸ‰ å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿæ¼”ç¤ºå®Œæˆ!")
print("=" * 50)

# é¢å¤–åŠŸèƒ½æ¼”ç¤º
print(f"\nğŸ“š ç³»ç»ŸåŠŸèƒ½æ¼”ç¤º:")

# æŸ¥è¯¢å•ä¸ªå­¦ç”Ÿæˆç»©
print("\n� æŸ¥è¯¢åŠŸèƒ½æ¼”ç¤º:")
student_to_check = "2024001"
if student_to_check in students:
    student_info = students[student_to_check]
    avg_score = get_student_average(student_to_check)
    print(f"å­¦ç”Ÿ {student_info['name']} (ID: {student_to_check}):")
    print(f"  å¹³å‡åˆ†: {avg_score:.1f}")
    for course, score in student_info['grades'].items():
        print(f"  {course}: {score}åˆ†")

# æŸ¥è¯¢è¯¾ç¨‹ä¿¡æ¯
print(f"\nğŸ“Š è¯¾ç¨‹åˆ†æ:")
course_to_analyze = "Pythonç¼–ç¨‹"
stats = get_course_statistics(course_to_analyze)
if stats:
    print(f"{course_to_analyze} è¯¾ç¨‹åˆ†æ:")
    print(f"  å‚ä¸å­¦ç”Ÿ: {stats['count']} äºº")
    print(f"  å¹³å‡åˆ†: {stats['average']:.1f}")
    print(f"  æœ€é«˜åˆ†: {stats['max']}")
    print(f"  æœ€ä½åˆ†: {stats['min']}")
    print(f"  åŠæ ¼ç‡: {stats['pass_rate']:.1f}%")

# æ•°æ®ç»Ÿè®¡
print(f"\nğŸ“ˆ æ•´ä½“ç»Ÿè®¡:")
total_students = len(students)
total_courses = len(courses)
total_grades = sum(len(student['grades']) for student in students.values())

print(f"  å­¦ç”Ÿæ€»æ•°: {total_students}")
print(f"  è¯¾ç¨‹æ€»æ•°: {total_courses}")
print(f"  æˆç»©è®°å½•æ€»æ•°: {total_grades}")

# æ‰¾å‡ºæœ€ä¼˜ç§€çš„å­¦ç”Ÿ
if students:
    best_student_id = None
    best_average = 0
    
    for student_id in students:
        avg = get_student_average(student_id)
        if avg and avg > best_average:
            best_average = avg
            best_student_id = student_id
    
    if best_student_id:
        best_student_name = students[best_student_id]['name']
        print(f"  æœ€ä¼˜ç§€å­¦ç”Ÿ: {best_student_name} (å¹³å‡åˆ†: {best_average:.1f})")

print(f"\nğŸ“š æœ¬é¡¹ç›®ä½¿ç”¨çš„çŸ¥è¯†ç‚¹:")
print("âœ… å­—å…¸å­˜å‚¨å¤æ‚æ•°æ®ç»“æ„")
print("âœ… åˆ—è¡¨ç®¡ç†é›†åˆæ•°æ®")
print("âœ… å‡½æ•°å°è£…åŠŸèƒ½æ¨¡å—")
print("âœ… å…¨å±€å˜é‡ç®¡ç†ç¨‹åºçŠ¶æ€")
print("âœ… forå¾ªç¯éå†æ•°æ®")
print("âœ… æ¡ä»¶è¯­å¥è¿›è¡Œé€»è¾‘åˆ¤æ–­")
print("âœ… å­—ç¬¦ä¸²æ ¼å¼åŒ–è¾“å‡º")
print("âœ… æ•°å­¦ç»Ÿè®¡è®¡ç®—")
print("âœ… å­—å…¸æ–¹æ³•ï¼škeys(), values(), items()")
print("âœ… åˆ—è¡¨æ¨å¯¼å¼è¿›è¡Œæ•°æ®è¿‡æ»¤")

print(f"\nğŸš€ æ­å–œå®Œæˆ Lesson-2 çš„å­¦ä¹ !")
print("ä¸‹èŠ‚è¯¾æˆ‘ä»¬å°†å­¦ä¹ é¢å‘å¯¹è±¡ç¼–ç¨‹ï¼Œè®©ç¨‹åºç»“æ„æ›´åŠ æ¸…æ™°å’Œå¼ºå¤§ï¼")
```
