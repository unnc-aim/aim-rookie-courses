### 1.1 列表(List) - 有序可变集合

列表是 Python 中最常用的数据结构之一，可以存储多个有序的元素，并且可以修改。


```python
# 列表的创建

print("=== 列表创建 ===")

# 创建列表的不同方式
robots = ["Hero", "Infantry", "Engineer"]  # 三种主要机器人类型
robot_hp = [600, 200, 250]                # 对应的血量
robot_damage = [10, 1, 0]                 # 对应的攻击伤害
mixed_list = ["Hero", 600, True, 10.5]    # 混合数据类型
empty_list = []
range_list = list(range(1, 6))             # [1, 2, 3, 4, 5]

print(f"机器人类型: {robots}")
print(f"机器人血量: {robot_hp}")
print(f"机器人伤害: {robot_damage}")
print(f"混合列表: {mixed_list}")
print(f"空列表: {empty_list}")
print(f"范围列表: {range_list}")
```

```python
# 列表访问和索引

robots = ["Hero", "Infantry", "Engineer"]
robot_hp = [600, 200, 250]
robot_damage = [10, 1, 0]

print("=== 列表访问 ===")
print(f"第一种机器人: {robots[0]} (血量: {robot_hp[0]}HP, 伤害: {robot_damage[0]}点)")
print(f"最后一种机器人: {robots[-1]} (血量: {robot_hp[-1]}HP, 伤害: {robot_damage[-1]}点)")
print(f"步兵机器人: {robots[1]} (血量: {robot_hp[1]}HP, 伤害: {robot_damage[1]}点)")
```

```python
# 列表切片

# [a:b:c]
# a -> 起始索引（包含）
# b -> 结束索引（不包含）
# c -> 步长（每隔几个项目取一个，默认是1）
# [0:] = [0:8:1] -> 从头开始到索引1（不包含1）
# [:4] = [0:4:1] -> 从头开始到索引4（不包含4）
# [4:] = [4:8:1] -> 从索引4开始到结尾
# [2:6] = [2:6:1] -> 从索引2开始到索引6（不包含6）
# [::2] = [0:8:2] -> 从头开始到结尾，每隔一个项目取一个
# [::-1] = [-1,-len(list)-1,-1] -> 反转列表

robots = ["Hero", "Infantry", "Engineer"]
robot_hp = [600, 200, 250]

print("=== 列表切片 ===")
print(f"前两种机器人: {robots[:2]}")
print(f"后两种机器人: {robots[-2:]}")
print(f"血量切片: {robot_hp[1:3]}")
print(f"反转列表: {robots[::-1]}")
```

```python
# 列表修改

robots = ["Hero", "Infantry", "Engineer"]

print("=== 列表修改 ===")
print(f"原始列表: {robots}")

robots[1] = "Infantry-Advanced"  # 修改单个元素
print(f"修改单个元素后: {robots}")

robots[0:2] = ["Hero-Pro", "Infantry-Elite"]  # 修改多个元素
print(f"修改多个元素后: {robots}")
```

```python
# 列表添加元素

battle_robots = ["Hero"]

print("=== 列表添加元素 ===")
print(f"初始战斗队列: {battle_robots}")

battle_robots.append("Infantry")           # 在末尾添加
print(f"append后: {battle_robots}")

battle_robots.insert(1, "Engineer")        # 在指定位置插入
print(f"insert后: {battle_robots}")

battle_robots.extend(["Infantry", "Infantry"])  # 添加多个元素
print(f"extend后: {battle_robots}")
```

```python
# 列表删除元素

battle_robots = ["Hero", "Infantry", "Engineer", "Infantry", "Infantry"]

print("=== 列表删除元素 ===")
print(f"初始列表: {battle_robots}")

removed_robot = battle_robots.pop()        # 删除并返回最后一个
print(f"pop()删除最后一个: {removed_robot}")
print(f"删除后: {battle_robots}")

battle_robots.remove("Engineer")           # 删除指定元素
print(f"remove('Engineer')后: {battle_robots}")

del battle_robots[1]                       # 删除指定位置
print(f"del battle_robots[1]后: {battle_robots}")
```

```python
# 列表信息查询

all_robot_hp = [600, 200, 250, 200, 600, 250]  # 多个机器人的血量

print("=== 列表信息 ===")
print(f"所有机器人血量: {all_robot_hp}")
print(f"机器人数量: {len(all_robot_hp)}")
print(f"最高血量: {max(all_robot_hp)}")
print(f"最低血量: {min(all_robot_hp)}")
print(f"总血量: {sum(all_robot_hp)}")
print(f"平均血量: {sum(all_robot_hp)/len(all_robot_hp):.1f}")
print(f"英雄机器人(600HP)数量: {all_robot_hp.count(600)}")
print(f"工程机器人血量(250HP)的位置: {all_robot_hp.index(250)}")
```

```python
# 列表排序

unsorted_hp = [250, 600, 200, 200, 600]

print("=== 列表排序 ===")
print(f"原始血量: {unsorted_hp}")

sorted_ascending = sorted(unsorted_hp)       # 不改变原列表，返回新列表
sorted_descending = sorted(unsorted_hp, reverse=True)

print(f"血量升序: {sorted_ascending}")
print(f"血量降序: {sorted_descending}")
print(f"原列表不变: {unsorted_hp}")

# 原地排序
unsorted_hp.sort(reverse=True)
print(f"原地降序排列后: {unsorted_hp}")
```

```python
# 列表推导式

hp_values = [600, 200, 250]

print("=== 列表推导式 ===")
hp_after_damage = [hp - 50 for hp in hp_values]  # 每个机器人受到50点伤害
high_hp_robots = [hp for hp in hp_values if hp > 250]  # 筛选高血量机器人

print(f"原始血量: {hp_values}")
print(f"受伤后血量: {hp_after_damage}")
print(f"高血量机器人: {high_hp_robots}")

# 复杂的列表推导式
numbers = [1, 2, 3, 4, 5]
squared_evens = [x**2 for x in numbers if x % 2 == 0]
print(f"偶数的平方: {squared_evens}")
```

### 1.2 字典(Dict) - 键值对映射

字典是 Python 中非常重要的数据结构，用于存储键值对，类似于现实中的字典。


```python
# 字典的创建

print("=== 字典创建 ===")

# 创建字典的不同方式
robot_config = {
    "name": "Hero",
    "type": "英雄机器人",
    "hp": 600,           # 英雄机器人血量厚
    "damage": 10,        # 大弹丸攻击力
    "weapon": "42mm大弹丸",
    "active": True
}

# 另一种创建方式
team_scores = dict(red=150, blue=120, green=80)

# 空字典
empty_dict = {}
empty_dict2 = dict()

print(f"英雄机器人配置: {robot_config}")
print(f"队伍得分: {team_scores}")
print(f"空字典: {empty_dict}")
```

> 注意：我们倾向于认为字典是一种无序的数据结构，但是 从 Python 3.7 开始，字典在语言层面保证了插入顺序的保持。也就是说，遍历字典时，会按照键值对被插入的先后顺序返回。
>
> 在 Python 3.6 及之前，字典的实现虽然在 CPython 中已经表现出“插入有序”的特性，但这只是实现细节，并没有在语言规范中保证。因此如果你需要编写兼容早期版本的代码，就不要依赖字典的顺序。
>
> 如果你的程序逻辑对顺序有严格要求（尤其是在旧版本 Python 中），可以使用 collections.OrderedDict，它在所有 Python 版本中都显式地保证顺序一致。

### 2.1 条件语句 - 让程序做决策

条件语句让程序能够根据不同的情况执行不同的代码，这是编程逻辑的核心。


```python
# 基本条件语句

print("=== 基本条件语句 ===")

# 简单if语句
robot_hp = 150

if robot_hp > 200:
    status = "状态优秀"
    action = "继续战斗"
elif robot_hp > 100:
    status = "状态良好"
    action = "小心应战"
elif robot_hp > 50:
    status = "状态一般"
    action = "寻找掩护"
else:
    status = "状态危急"
    action = "立即撤退"

print(f"机器人血量: {robot_hp}")
print(f"状态评估: {status}")
print(f"建议行动: {action}")
```

```python
# 复杂条件判断

print("=== 复杂条件判断 ===")

# 机器人状态参数
hp = 120
armor = 40
ammo = 60
energy = 65

print(f"分析参数: HP={hp}, 护甲={armor}, 弹药={ammo}, 能量={energy}")

# 多条件组合判断
if hp <= 0:
    status = "机器人已损坏"
    action = "需要维修"
elif hp > 150 and armor > 50 and ammo > 100 and energy > 70:
    status = "战斗状态极佳"
    action = "主动出击"
elif hp > 100 and (armor > 30 or ammo > 50) and energy > 50:
    status = "战斗状态良好"
    action = "谨慎推进"
elif hp > 50 and energy > 30:
    if ammo <= 20:
        status = "弹药不足"
        action = "补充弹药"
    elif armor <= 20:
        status = "护甲受损"
        action = "维修护甲"
    else:
        status = "状态一般"
        action = "保守作战"
else:
    status = "状态危急"
    action = "立即撤退"

print(f"状态评估: {status}")
print(f"建议行动: {action}")
```

```python
# 三元运算符（条件表达式）

print("=== 三元运算符 ===")

# 基本三元运算符
robot_energy = 75
battery_status = "充足" if robot_energy > 50 else "不足"
print(f"能量: {robot_energy}%, 电池状态: {battery_status}")

# 机器人状态检查
robot_hp = 180
health_status = "健康" if robot_hp > 100 else "受伤"
print(f"血量: {robot_hp}, 健康状态: {health_status}")

# 弹药检查
ammo_count = 25
ammo_status = "充足" if ammo_count > 30 else "不足"
print(f"弹药: {ammo_count}发, 弹药状态: {ammo_status}")
```

```python
# 链式比较和多重三元运算符

print("=== 链式比较 ===")

# 链式比较
temperature = 25
print(f"温度: {temperature}°C")

if 20 <= temperature <= 30:
    comfort_level = "舒适"
elif temperature > 30:
    comfort_level = "过热"
else:
    comfort_level = "过冷"

print(f"舒适度: {comfort_level}")

# 多重三元运算符
robot_status = (
    "过热" if temperature > 30 else
    "舒适" if 20 <= temperature <= 30 else
    "过冷"
)
print(f"机器人状态: {robot_status}")
```

```python
# 机器人指令处理

print("=== 机器人指令处理 ===")

# 不同指令的处理
commands = ["attack", "defend", "move_forward", "target_enemy1", "invalid_cmd"]

for command in commands:
    print(f"\n处理指令: '{command}'")
    
    # 使用if-elif处理不同指令
    if command == "attack":
        result = "执行攻击指令"
    elif command == "defend":
        result = "进入防御模式"
    elif command == "retreat":
        result = "开始撤退"
    elif command in ["move_forward", "move_backward", "move_left", "move_right"]:
        result = f"执行移动指令: {command}"
    elif command.startswith("target_"):
        target = command.split("_")[1]
        result = f"锁定目标: {target}"
    else:
        result = "未知指令，请重新输入"
    
    print(f"结果: {result}")
```

```python
# 队伍表现评估

print("=== 队伍表现评估 ===")

# 队伍得分数据
team_scores = [88, 92, 85, 90, 87]

print(f"队伍分数: {team_scores}")

# 检查是否有分数数据
if not team_scores:  # 检查空列表
    print("没有分数数据")
    performance = "未知"
    average = 0.0
    score_range = 0
    consistency = "未知"
else:
    # 计算统计数据
    avg_score = sum(team_scores) / len(team_scores)
    max_score = max(team_scores)
    min_score = min(team_scores)
    
    # 使用常量提高代码可读性
    EXCELLENT_THRESHOLD = 90
    GOOD_THRESHOLD = 75
    AVERAGE_THRESHOLD = 60
    
    # 评估表现等级
    if avg_score >= EXCELLENT_THRESHOLD:
        performance = "优秀"
    elif avg_score >= GOOD_THRESHOLD:
        performance = "良好"
    elif avg_score >= AVERAGE_THRESHOLD:
        performance = "一般"
    else:
        performance = "需要改进"
    
    # 额外分析
    score_range = max_score - min_score
    consistency = "稳定" if score_range <= 20 else "不稳定"
    average = round(avg_score, 1)

print(f"表现评估: {performance}")
print(f"平均分: {average}")
print(f"分数波动: {score_range} (状态: {consistency})")
```

### 2.2 循环语句 - 重复执行任务

循环让程序能够重复执行相同的操作，这在处理大量数据或重复性任务时非常有用。


```python
# 基本for循环
print("=== for循环 - 遍历序列 ===")

robots = ["Hero", "Infantry", "Engineer", "Drone"]
print("检查机器人状态:")
for robot in robots:
    print(f"  正在检查 {robot}...")
```

```python
# 遍历索引和值
robots = ["Hero", "Infantry", "Engineer", "Drone"]
print(f"带索引的遍历:")
for index, robot in enumerate(robots):
    print(f"  {index + 1}. {robot}")
```

```python
# 多个测试案例

print("=== 多个测试案例 ===")

# 测试不同状态的机器人
test_cases = [
    (200, 80, 150, 90, "最佳状态"),  # 最佳状态
    (120, 40, 60, 65, "良好状态"),   # 良好状态
    (80, 15, 30, 45, "一般状态"),    # 一般状态
    (30, 5, 10, 20, "危急状态"),     # 危急状态
    (0, 0, 0, 0, "损坏状态")         # 损坏状态
]

for i, (hp, armor, ammo, energy, description) in enumerate(test_cases, 1):
    print(f"\n测试{i} - {description}:")
    print(f"  参数: HP={hp}, 护甲={armor}, 弹药={ammo}, 能量={energy}")
    
    # 状态判断逻辑
    if hp <= 0:
        status = "机器人已损坏"
        action = "需要维修"
    elif hp > 150 and armor > 50 and ammo > 100 and energy > 70:
        status = "战斗状态极佳"
        action = "主动出击"
    elif hp > 100 and (armor > 30 or ammo > 50) and energy > 50:
        status = "战斗状态良好"
        action = "谨慎推进"
    elif hp > 50 and energy > 30:
        if ammo <= 20:
            status = "弹药不足"
            action = "补充弹药"
        elif armor <= 20:
            status = "护甲受损"
            action = "维修护甲"
        else:
            status = "状态一般"
            action = "保守作战"
    else:
        status = "状态危急"
        action = "立即撤退"
    
    print(f"  结果: {status} -> {action}")
```

```python
# 用户输入验证
robot_id = None
attempts = 0
max_attempts = 5

while attempts < max_attempts and robot_id is None:
    user_input = input("请输入机器人ID (1-6): ")
    if user_input.isdigit() and 1 <= int(user_input) <= 6:
        robot_id = int(user_input)
    else:
        print("输入无效，请重新输入")
        attempts += 1

if robot_id is None:
    print("多次输入无效，操作终止。")
else:
    print(f"选择的机器人ID: {robot_id}")
```

> 注意：非必要不要在程序中使用 `while True` 这种无限循环的写法，因为如果没有合适的跳出条件，程序会一直运行下去，可能导致系统资源耗尽或程序崩溃。
>
> 除非你非常确定需要一个无限循环，并且有明确的跳出机制，否则建议使用有限次数的循环或者其他更安全的循环结构。

```python
# break 和 continue 用例

# for循环中的break和continue
commands = ["attack", "move_forward", "defend", "invalid_cmd", "retreat", "attack"]

print("\nfor循环示例:")
for cmd in commands:
    if cmd == "invalid_cmd":
        print(f"遇到无效指令: {cmd}，跳过（continue）")
        continue
    print(f"处理指令: {cmd}")
    if cmd == "retreat":
        print("收到撤退指令，立即停止处理（break）")
        break
```

```python
# while循环中的break和continue
print("\nwhile循环示例:")
attempts = 1
max_attempts = 5
while attempts <= max_attempts:
    user_input = input(f"第{attempts}次输入机器人ID (1-6): ")
    if not user_input.isdigit():
        print("输入不是数字，跳过本次（continue）")
        attempts += 1
        continue
    robot_id = int(user_input)
    if 1 <= robot_id <= 6:
        print(f"已选择机器人ID: {robot_id}")
        break
    else:
        print("ID无效，请重新输入")
        attempts += 1
else:
    print("多次输入无效，操作终止。")
```

### 3.1 函数定义和参数

函数是代码复用的重要工具，让我们能够将复杂的任务分解为简单的、可重复使用的模块。


```python
# 基本函数定义
def calculate_damage(attack_power, armor_defense):
    """
    计算实际伤害值

    Args:
        attack_power (int): 攻击力
        armor_defense (int): 护甲防御力

    Returns:
        int: 实际造成的伤害
    """
    damage = max(1, attack_power - armor_defense)  # 至少造成1点伤害
    return damage


# 测试基本函数
damage1 = calculate_damage(50, 20)
damage2 = calculate_damage(30, 35)
print(f"攻击力50 vs 护甲20 = {damage1}点伤害")
print(f"攻击力30 vs 护甲35 = {damage2}点伤害")
```

```python
# 默认参数
def create_robot(name, robot_type="Infantry", hp=200, speed=3.0):
    """
    创建机器人对象

    Args:
        name (str): 机器人名称
        robot_type (str): 机器人类型，默认为Infantry
        hp (int): 生命值，默认为200
        speed (float): 移动速度，默认为3.0

    Returns:
        dict: 机器人信息字典
    """
    robot = {
        "name": name,
        "type": robot_type,
        "hp": hp,
        "max_hp": hp,
        "speed": speed,
        "status": "待命"
    }
    return robot


# 测试默认参数
print(f"\\n=== 默认参数测试 ===")
robot1 = create_robot("普通步兵")
robot2 = create_robot("英雄机器人", "Hero", 400, 2.5)
robot3 = create_robot(name="工程机器人", robot_type="Engineer", speed=4)

print(f"机器人1: {robot1}")
print(f"机器人2: {robot2}")
print(f"机器人3: {robot3}")
```

```python
# 可变参数 (*args)
def calculate_total_score(*scores):
    """
    计算总分和平均分

    Args:
        *scores: 可变数量的分数

    Returns:
        tuple: (总分, 平均分, 分数个数)
    """
    if not scores:
        return 0, 0, 0

    total = sum(scores)
    average = total / len(scores)
    return total, average, len(scores)


# 测试可变参数
print(f"\\n=== 可变参数测试 ===")
result1 = calculate_total_score(85, 92, 78, 90)
result2 = calculate_total_score(95, 88, 76, 89, 91, 84)

print(f"4个分数: 总分={result1[0]}, 平均={result1[1]:.1f}, 个数={result1[2]}")
print(f"6个分数: 总分={result2[0]}, 平均={result2[1]:.1f}, 个数={result2[2]}")
```

```python
# 关键字参数 (**kwargs)
def create_team(**members):
    """
    创建队伍

    Args:
        **members: 队员信息的关键字参数

    Returns:
        dict: 队伍信息
    """
    team = {
        "members": members,
        "size": len(members),
        "captain": members.get("captain", "未指定")
    }
    return team


# 测试关键字参数
print(f"\\n=== 关键字参数测试 ===")
team1 = create_team(captain="张三", programmer="李四", operator="王五")
team2 = create_team(leader="赵六", member1="孙七", member2="周八", member3="吴九")

print(f"队伍1: {team1}")
print(f"队伍2: {team2}")
```

```python
# 混合参数类型
def advanced_robot_config(name, robot_type, *weapons, hp=200, **special_abilities):
    """
    高级机器人配置函数

    Args:
        name (str): 机器人名称
        robot_type (str): 机器人类型
        *weapons: 武器列表
        hp (int): 生命值
        **special_abilities: 特殊能力

    Returns:
        dict: 完整的机器人配置
    """
    config = {
        "name": name,
        "type": robot_type,
        "hp": hp,
        "weapons": list(weapons),
        "abilities": special_abilities
    }
    return config


# 测试混合参数
print(f"\\n=== 混合参数测试 ===")
hero_config = advanced_robot_config(
    "超级英雄", "Hero",
    "17mm机枪", "42mm大炮",
    hp=400,
    shield=True,
    boost_speed=True,
    special_attack="能量炮"
)

print(f"英雄配置: {hero_config}")
```

```python
# 函数作为参数
def apply_strategy(robots, strategy_func):
    """
    对机器人列表应用策略函数

    Args:
        robots (list): 机器人列表
        strategy_func (function): 策略函数

    Returns:
        list: 应用策略后的结果
    """
    return [strategy_func(robot) for robot in robots]


def attack_strategy(robot):
    """攻击策略"""
    return f"{robot['name']} 执行攻击任务"


def defense_strategy(robot):
    """防御策略"""
    return f"{robot['name']} 执行防御任务"


# 测试函数作为参数
robots = [
    {"name": "Hero-1", "type": "Hero"},
    {"name": "Infantry-1", "type": "Infantry"},
    {"name": "Engineer-1", "type": "Engineer"}
]

print(f"\\n=== 函数作为参数测试 ===")
attack_results = apply_strategy(robots, attack_strategy)
defense_results = apply_strategy(robots, defense_strategy)

print("攻击策略结果:")
for result in attack_results:
    print(f"  {result}")

print("防御策略结果:")
for result in defense_results:
    print(f"  {result}")
```

```python
# 闭包和装饰器预览
def create_robot_factory(robot_type):
    """
    创建机器人工厂函数（闭包示例）

    Args:
        robot_type (str): 机器人类型

    Returns:
        function: 创建特定类型机器人的函数
    """
    def create_specific_robot(name, hp=200):
        return {
            "name": name,
            "type": robot_type,
            "hp": hp
        }
    return create_specific_robot


# 创建专门的工厂函数
hero_factory = create_robot_factory("Hero")
infantry_factory = create_robot_factory("Infantry")

# 使用工厂函数
hero1 = hero_factory("Hero-Alpha", 400)
infantry1 = infantry_factory("Infantry-Beta", 200)

print(f"英雄机器人: {hero1}")
print(f"步兵机器人: {infantry1}")
```

```python
# 递归函数示例
def calculate_fibonacci(n):
    """
    计算斐波那契数列

    Args:
        n (int): 序列位置

    Returns:
        int: 斐波那契数值
    """
    if n <= 1:
        return n
    return calculate_fibonacci(n-1) + calculate_fibonacci(n-2)

# When calling calculate_fibonacci(3):
#
# cf(3)
# -> cf(2) + cf(1)
# -> (cf(1) + cf(0)) + 1
# -> 1 + 1
# -> 2


print(f"\\n=== 递归函数示例 ===")
fib_sequence = [calculate_fibonacci(i) for i in range(10)]
print(f"斐波那契数列前10项: {fib_sequence}")

print("\\n✅ 函数进阶学习完成！")
```

```python
# Lesson-2 综合实践项目：学生成绩管理系统

print("🎓 学生成绩管理系统")
print("=" * 50)

# 使用全局变量存储数据
students = {}  # 存储学生信息
courses = []   # 存储课程列表

def add_course(course_name):
    """添加课程"""
    global courses
    if course_name not in courses:
        courses.append(course_name)
        print(f"✅ 已添加课程: {course_name}")
    else:
        print(f"❌ 课程 {course_name} 已存在")

def add_student(student_id, name):
    """添加学生"""
    global students
    if student_id not in students:
        students[student_id] = {
            "name": name,
            "grades": {}
        }
        print(f"✅ 已添加学生: {name} (ID: {student_id})")
    else:
        print(f"❌ 学生ID {student_id} 已存在")

def add_grade(student_id, course, score):
    """添加成绩"""
    global students, courses
    if student_id in students:
        if course in courses:
            students[student_id]["grades"][course] = score
            print(f"✅ 已录入 {students[student_id]['name']} 的 {course} 成绩: {score}")
        else:
            print(f"❌ 课程 {course} 不存在，请先添加课程")
    else:
        print(f"❌ 学生ID {student_id} 不存在")

def get_student_average(student_id):
    """计算学生平均分"""
    global students
    if student_id in students:
        grades = list(students[student_id]["grades"].values())
        if grades:
            return sum(grades) / len(grades)
        else:
            return 0
    return None

def get_course_statistics(course):
    """获取课程统计信息"""
    global students
    scores = []
    for student in students.values():
        if course in student["grades"]:
            scores.append(student["grades"][course])
    
    if scores:
        return {
            "count": len(scores),
            "average": sum(scores) / len(scores),
            "max": max(scores),
            "min": min(scores),
            "pass_rate": len([s for s in scores if s >= 60]) / len(scores) * 100
        }
    return None

def display_all_students():
    """显示所有学生信息"""
    global students
    print(f"\n📊 学生信息总览:")
    print("-" * 60)
    
    for student_id, info in students.items():
        name = info["name"]
        grades = info["grades"]
        avg = get_student_average(student_id)
        
        print(f"学号: {student_id} | 姓名: {name} | 平均分: {avg:.1f}")
        
        if grades:
            for course, score in grades.items():
                status = "✅ 及格" if score >= 60 else "❌ 不及格"
                print(f"  └─ {course}: {score}分 {status}")
        else:
            print("  └─ 暂无成绩记录")
        print()

def display_course_statistics():
    """显示课程统计"""
    global courses
    print(f"\n📈 课程统计信息:")
    print("-" * 60)
    
    for course in courses:
        stats = get_course_statistics(course)
        if stats:
            print(f"课程: {course}")
            print(f"  参与人数: {stats['count']}")
            print(f"  平均分: {stats['average']:.1f}")
            print(f"  最高分: {stats['max']}")
            print(f"  最低分: {stats['min']}")
            print(f"  及格率: {stats['pass_rate']:.1f}%")
            print()
        else:
            print(f"课程: {course} (暂无成绩数据)\n")

def reset_system():
    """重置系统数据"""
    global students, courses
    students = {}
    courses = []
    print("🔄 系统数据已重置")

def demo_grade_system():
    """演示成绩管理系统"""
    
    # 重置系统
    reset_system()
    
    # 添加课程
    print("1️⃣ 添加课程")
    course_list = ["Python编程", "数据结构", "算法设计", "机器学习"]
    for course in course_list:
        add_course(course)
    
    print(f"\n2️⃣ 添加学生")
    # 添加学生
    student_list = [
        ("2024001", "张三"),
        ("2024002", "李四"), 
        ("2024003", "王五"),
        ("2024004", "赵六"),
        ("2024005", "孙七")
    ]
    
    for student_id, name in student_list:
        add_student(student_id, name)
    
    print(f"\n3️⃣ 录入成绩")
    # 录入成绩
    grades_data = [
        ("2024001", "Python编程", 95),
        ("2024001", "数据结构", 88),
        ("2024001", "算法设计", 92),
        ("2024002", "Python编程", 78),
        ("2024002", "数据结构", 85),
        ("2024002", "机器学习", 90),
        ("2024003", "Python编程", 56),
        ("2024003", "数据结构", 72),
        ("2024003", "算法设计", 68),
        ("2024004", "Python编程", 91),
        ("2024004", "机器学习", 87),
        ("2024005", "Python编程", 83),
        ("2024005", "数据结构", 79),
        ("2024005", "算法设计", 85),
        ("2024005", "机器学习", 92)
    ]
    
    for student_id, course, score in grades_data:
        add_grade(student_id, course, score)
    
    # 显示结果
    display_all_students()
    display_course_statistics()

# 运行演示
demo_grade_system()

print("\n" + "=" * 50)
print("🎉 学生成绩管理系统演示完成!")
print("=" * 50)

# 额外功能演示
print(f"\n📚 系统功能演示:")

# 查询单个学生成绩
print("\n� 查询功能演示:")
student_to_check = "2024001"
if student_to_check in students:
    student_info = students[student_to_check]
    avg_score = get_student_average(student_to_check)
    print(f"学生 {student_info['name']} (ID: {student_to_check}):")
    print(f"  平均分: {avg_score:.1f}")
    for course, score in student_info['grades'].items():
        print(f"  {course}: {score}分")

# 查询课程信息
print(f"\n📊 课程分析:")
course_to_analyze = "Python编程"
stats = get_course_statistics(course_to_analyze)
if stats:
    print(f"{course_to_analyze} 课程分析:")
    print(f"  参与学生: {stats['count']} 人")
    print(f"  平均分: {stats['average']:.1f}")
    print(f"  最高分: {stats['max']}")
    print(f"  最低分: {stats['min']}")
    print(f"  及格率: {stats['pass_rate']:.1f}%")

# 数据统计
print(f"\n📈 整体统计:")
total_students = len(students)
total_courses = len(courses)
total_grades = sum(len(student['grades']) for student in students.values())

print(f"  学生总数: {total_students}")
print(f"  课程总数: {total_courses}")
print(f"  成绩记录总数: {total_grades}")

# 找出最优秀的学生
if students:
    best_student_id = None
    best_average = 0
    
    for student_id in students:
        avg = get_student_average(student_id)
        if avg and avg > best_average:
            best_average = avg
            best_student_id = student_id
    
    if best_student_id:
        best_student_name = students[best_student_id]['name']
        print(f"  最优秀学生: {best_student_name} (平均分: {best_average:.1f})")
```

```python
def degrees_to_radians(degrees):
    """将角度转换为弧度"""
    return degrees * math.pi / 180


print(degrees_to_radians(180))  # 应输出 3.14159 (π)
print(degrees_to_radians(90))   # 应输出 1.5708 (π/2)
```

```python
def triangle_area(base, height):
    """计算三角形面积"""
    return 0.5 * base * height


print(triangle_area(5, 10))  # 应输出 25.0
print(triangle_area(8, 6))   # 应输出 24.0
```

```python
def is_prime(n):
    """判断一个数是否为质数"""
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


print(is_prime(11))  # 应输出 True
print(is_prime(4))   # 应输出 False
print(is_prime(2))  # 应输出 True
```
