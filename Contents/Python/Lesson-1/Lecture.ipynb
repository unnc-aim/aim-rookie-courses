```python
def simple_calculator():
    """
    RoboMaster ç®€å•è®¡ç®—å™¨
    æ”¯æŒåŸºæœ¬çš„å››åˆ™è¿ç®—å’Œé”™è¯¯å¤„ç†
    """
    print("=" * 50)
    print("ğŸ¤– RoboMaster ç®€å•è®¡ç®—å™¨ ğŸ¤–")
    print("=" * 50)
    print("æ”¯æŒçš„è¿ç®—ç¬¦: +, -, *, /, //, %, **")
    print("è¾“å…¥ 'quit' é€€å‡ºç¨‹åº")
    print("=" * 50)
    
    while True:
        try:
            # æ¨¡æ‹Ÿç”¨æˆ·è¾“å…¥ï¼ˆåœ¨å®é™…ä½¿ç”¨ä¸­ä¼šæ˜¯çœŸå®è¾“å…¥ï¼‰
            # è¿™é‡Œæˆ‘ä»¬æ¼”ç¤ºå‡ ä¸ªè®¡ç®—ä¾‹å­
            
            # ç¤ºä¾‹è®¡ç®—1: åŸºæœ¬åŠ æ³•
            print("\\nç¤ºä¾‹1: åŸºæœ¬åŠ æ³•")
            print("è¯·è¾“å…¥ç¬¬ä¸€ä¸ªæ•°å­—: ", end="")
            num1 = 10.5
            print(num1)
            
            print("è¯·è¾“å…¥è¿ç®—ç¬¦ (+, -, *, /, //, %, **): ", end="")
            operator = "+"
            print(operator)
            
            print("è¯·è¾“å…¥ç¬¬äºŒä¸ªæ•°å­—: ", end="")
            num2 = 5.2
            print(num2)
            
            result = perform_calculation(num1, operator, num2)
            if result is not None:
                print(f"è®¡ç®—ç»“æœ: {num1} {operator} {num2} = {result}")
            
            # ç¤ºä¾‹è®¡ç®—2: é™¤æ³•
            print("\\nç¤ºä¾‹2: é™¤æ³•è¿ç®—")
            print("è¯·è¾“å…¥ç¬¬ä¸€ä¸ªæ•°å­—: ", end="")
            num1 = 15
            print(num1)
            
            print("è¯·è¾“å…¥è¿ç®—ç¬¦ (+, -, *, /, //, %, **): ", end="")
            operator = "/"
            print(operator)
            
            print("è¯·è¾“å…¥ç¬¬äºŒä¸ªæ•°å­—: ", end="")
            num2 = 4
            print(num2)
            
            result = perform_calculation(num1, operator, num2)
            if result is not None:
                print(f"è®¡ç®—ç»“æœ: {num1} {operator} {num2} = {result}")
            
            # ç¤ºä¾‹è®¡ç®—3: é™¤é›¶é”™è¯¯
            print("\\nç¤ºä¾‹3: é™¤é›¶é”™è¯¯å¤„ç†")
            print("è¯·è¾“å…¥ç¬¬ä¸€ä¸ªæ•°å­—: ", end="")
            num1 = 10
            print(num1)
            
            print("è¯·è¾“å…¥è¿ç®—ç¬¦ (+, -, *, /, //, %, **): ", end="")
            operator = "/"
            print(operator)
            
            print("è¯·è¾“å…¥ç¬¬äºŒä¸ªæ•°å­—: ", end="")
            num2 = 0
            print(num2)
            
            result = perform_calculation(num1, operator, num2)
            if result is not None:
                print(f"è®¡ç®—ç»“æœ: {num1} {operator} {num2} = {result}")
            
            # ç¤ºä¾‹è®¡ç®—4: å¹‚è¿ç®—
            print("\\nç¤ºä¾‹4: å¹‚è¿ç®—")
            print("è¯·è¾“å…¥ç¬¬ä¸€ä¸ªæ•°å­—: ", end="")
            num1 = 2
            print(num1)
            
            print("è¯·è¾“å…¥è¿ç®—ç¬¦ (+, -, *, /, //, %, **): ", end="")
            operator = "**"
            print(operator)
            
            print("è¯·è¾“å…¥ç¬¬äºŒä¸ªæ•°å­—: ", end="")
            num2 = 10
            print(num2)
            
            result = perform_calculation(num1, operator, num2)
            if result is not None:
                print(f"è®¡ç®—ç»“æœ: {num1} {operator} {num2} = {result}")
            
            # ç¤ºä¾‹è®¡ç®—5: æ— æ•ˆè¿ç®—ç¬¦
            print("\\nç¤ºä¾‹5: æ— æ•ˆè¿ç®—ç¬¦å¤„ç†")
            print("è¯·è¾“å…¥ç¬¬ä¸€ä¸ªæ•°å­—: ", end="")
            num1 = 8
            print(num1)
            
            print("è¯·è¾“å…¥è¿ç®—ç¬¦ (+, -, *, /, //, %, **): ", end="")
            operator = "&"  # æ— æ•ˆè¿ç®—ç¬¦
            print(operator)
            
            print("è¯·è¾“å…¥ç¬¬äºŒä¸ªæ•°å­—: ", end="")
            num2 = 3
            print(num2)
            
            result = perform_calculation(num1, operator, num2)
            if result is not None:
                print(f"è®¡ç®—ç»“æœ: {num1} {operator} {num2} = {result}")
            
            break  # æ¼”ç¤ºç»“æŸ
            
        except KeyboardInterrupt:
            print("\\n\\nğŸ‘‹ æ„Ÿè°¢ä½¿ç”¨ RoboMaster è®¡ç®—å™¨!")
            break
        except Exception as e:
            print(f"å‘ç”Ÿæ„å¤–é”™è¯¯: {e}")

def perform_calculation(num1, operator, num2):
    """
    æ‰§è¡Œå…·ä½“çš„è®¡ç®—æ“ä½œ
    
    Args:
        num1: ç¬¬ä¸€ä¸ªæ•°å­—
        operator: è¿ç®—ç¬¦
        num2: ç¬¬äºŒä¸ªæ•°å­—
    
    Returns:
        è®¡ç®—ç»“æœï¼Œå¦‚æœå‡ºé”™è¿”å›None
    """
    try:
        if operator == '+':
            return num1 + num2
        elif operator == '-':
            return num1 - num2
        elif operator == '*':
            return num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("âŒ é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶!")
                return None
            return num1 / num2
        elif operator == '//':
            if num2 == 0:
                print("âŒ é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶!")
                return None
            return num1 // num2
        elif operator == '%':
            if num2 == 0:
                print("âŒ é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶!")
                return None
            return num1 % num2
        elif operator == '**':
            return num1 ** num2
        else:
            print(f"âŒ é”™è¯¯ï¼šä¸æ”¯æŒçš„è¿ç®—ç¬¦ '{operator}'!")
            print("æ”¯æŒçš„è¿ç®—ç¬¦: +, -, *, /, //, %, **")
            return None
            
    except OverflowError:
        print("âŒ é”™è¯¯ï¼šè®¡ç®—ç»“æœè¿‡å¤§!")
        return None
    except Exception as e:
        print(f"âŒ è®¡ç®—é”™è¯¯: {e}")
        return None

# è¿è¡Œè®¡ç®—å™¨æ¼”ç¤º
simple_calculator()

print("\\n" + "=" * 50)
print("ğŸ“š å®è·µé¡¹ç›®æ€»ç»“")
print("=" * 50)
print("è¿™ä¸ªè®¡ç®—å™¨é¡¹ç›®å±•ç¤ºäº†ä»¥ä¸‹ç¼–ç¨‹æ¦‚å¿µ:")
print("1. å‡½æ•°å®šä¹‰å’Œè°ƒç”¨")
print("2. ç”¨æˆ·è¾“å…¥å¤„ç†")
print("3. æ¡ä»¶è¯­å¥ (if-elif-else)")
print("4. å¼‚å¸¸å¤„ç† (try-except)")
print("5. å¾ªç¯æ§åˆ¶ (while)")
print("6. å­—ç¬¦ä¸²æ ¼å¼åŒ–")
print("7. åŸºæœ¬çš„ç¨‹åºè®¾è®¡æ¨¡å¼")
print("=" * 50)
```
