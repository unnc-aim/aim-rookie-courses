### 2.1 å°è£… (Encapsulation)

å°è£…æ˜¯å°†æ•°æ®å’Œæ“ä½œæ•°æ®çš„æ–¹æ³•ç»‘å®šåœ¨ä¸€èµ·ï¼Œå¹¶æ§åˆ¶å¤–éƒ¨è®¿é—®çš„ç‰¹æ€§ã€‚

```python
# å°è£… - å±æ€§ä¿æŠ¤å’Œè®¿é—®æ§åˆ¶

class Robot:
    """æ¼”ç¤ºå°è£…çš„æ­¥å…µæœºå™¨äººç±»"""

    def __init__(self, name, hp):
        self.name = name          # å…¬å¼€å±æ€§
        self._hp = hp            # å—ä¿æŠ¤å±æ€§ï¼ˆçº¦å®šç§æœ‰ï¼‰
        self._max_hp = hp
        self.__serial_number = f"RB{id(self)}"  # ç§æœ‰å±æ€§

    # @property ä¿®é¥°ç¬¦å°†æ–¹æ³•å˜ä¸ºå±æ€§ï¼Œå¯ç›´æ¥é€šè¿‡ robot.hp è®¿é—®è€Œä¸æ˜¯ robot.hp()
    # è¿™æ ·å¯ä»¥ä¿æŠ¤ _hpï¼Œä¸å…è®¸å¤–éƒ¨ç›´æ¥èµ‹å€¼ï¼Œå¿…é¡»é€šè¿‡ setter æ§åˆ¶
    # ä½“ç°å°è£…æ€æƒ³ï¼šå¯¹å¤–åªæš´éœ²å®‰å…¨çš„æ¥å£ï¼Œå†…éƒ¨ç»†èŠ‚éšè—
    @property
    def hp(self):
        """
        è¡€é‡å±æ€§ï¼ˆåªè¯»ï¼‰

        """
        return self._hp

    # @hp.setter ä¿®é¥°ç¬¦ç”¨äºå®šä¹‰ hp å±æ€§çš„â€œè®¾ç½®å™¨â€æ–¹æ³•
    # è¿™æ ·å¯ä»¥é€šè¿‡ robot.hp = xxx ç›´æ¥èµ‹å€¼ï¼Œå¹¶è‡ªåŠ¨è§¦å‘æ­¤æ–¹æ³•
    # èµ‹å€¼æ—¶ä¼šè‡ªåŠ¨æ£€æŸ¥èŒƒå›´ï¼Œé˜²æ­¢è¡€é‡è¶Šç•Œï¼Œä½“ç°å°è£…å’Œå®‰å…¨æ€§
    @hp.setter
    def hp(self, value):
        """è®¾ç½®è¡€é‡ï¼Œè‡ªåŠ¨é™åˆ¶èŒƒå›´"""
        if value < 0:
            self._hp = 0
        elif value > self._max_hp:
            self._hp = self._max_hp
        else:
            self._hp = value
        print(f"{self.name} è¡€é‡è®¾ç½®ä¸º: {self._hp}")
        print(f"å½“å‰è¡€é‡ç™¾åˆ†æ¯”: {self.calculate_hp_percentage(self._hp, self._max_hp):.2f}%")

    @property
    def max_hp(self):
        """æœ€å¤§è¡€é‡ï¼ˆåªè¯»ï¼‰"""
        return self._max_hp

    # @staticmethod è£…é¥°å™¨å°†æ–¹æ³•å˜ä¸ºé™æ€æ–¹æ³•
    # é™æ€æ–¹æ³•ä¸ä¾èµ–äºå®ä¾‹æˆ–ç±»ï¼Œå¯ä»¥ç›´æ¥é€šè¿‡ç±»åè°ƒç”¨
    # é€‚åˆå·¥å…·å‡½æ•°ï¼Œä½“ç°å°è£…å’Œç»„ç»‡ä»£ç çš„æ€æƒ³
    @staticmethod
    def calculate_hp_percentage(current, maximum):
        """è®¡ç®—è¡€é‡ç™¾åˆ†æ¯”"""
        if maximum == 0:
            return 0
        return (current / maximum) * 100

    def get_serial_number(self):
        """è·å–åºåˆ—å·ï¼ˆé—´æ¥è®¿é—®ç§æœ‰å±æ€§ï¼‰"""
        return self.__serial_number

    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        old_hp = self._hp
        self.hp = self._hp - damage  # é€šè¿‡setterè®¾ç½®
        print(f"{self.name} å—åˆ° {damage} ä¼¤å®³: {old_hp} -> {self._hp}")


# æµ‹è¯•å°è£…
robot = Robot("Hero-1", 100)

# æ­£å¸¸è®¿é—®
print(f"æœºå™¨äººåç§°: {robot.name}")
print(f"æœºå™¨äººè¡€é‡: {robot.hp}")
print(f"æœ€å¤§è¡€é‡: {robot.max_hp}")
print(f"åºåˆ—å·: {robot.get_serial_number()}")

# é€šè¿‡setterä¿®æ”¹è¡€é‡
robot.hp = 150  # ä¼šè¢«é™åˆ¶åœ¨max_hpä»¥å†…
robot.hp = -10  # ä¼šè¢«é™åˆ¶ä¸º0

# å—åˆ°ä¼¤å®³
robot.hp = 100  # é‡ç½®è¡€é‡
robot.take_damage(30)

print(Robot.calculate_hp_percentage(50, 200))  # 25.0
```

### 2.2 ç»§æ‰¿ (Inheritance)

ç»§æ‰¿æ˜¯ä»ç°æœ‰ç±»åˆ›å»ºæ–°ç±»çš„æœºåˆ¶ï¼Œå­ç±»å¯ä»¥ç»§æ‰¿çˆ¶ç±»çš„å±æ€§å’Œæ–¹æ³•ã€‚

```python
# ç»§æ‰¿ - åŸºç±»å’Œæ´¾ç”Ÿç±»


class StandardRobot:
    """æœºå™¨äººåŸºç±»ï¼Œä¹Ÿç”¨ä½œæ­¥å…µæœºå™¨äºº"""

    def __init__(self, name, hp):
        self.name = name
        self.hp = hp
        self.position = (0, 0)

    def move(self, x, y):
        """åŸºç¡€ç§»åŠ¨æ–¹æ³•"""
        self.position = (x, y)
        print(f"{self.name} ç§»åŠ¨åˆ° ({x}, {y})")

    def attack(self, target):
        """åŸºç¡€æ”»å‡»æ–¹æ³•"""
        damage = 5  # åŸºç¡€ä¼¤å®³
        target.hp -= damage
        print(f"{self.name} åŸºç¡€æ”»å‡» {target.name}ï¼Œé€ æˆ {damage} ç‚¹ä¼¤å®³")


class HeroRobot(StandardRobot):
    """è‹±é›„æœºå™¨äººç±»ï¼Œç»§æ‰¿è‡ªStandardRobot"""

    def __init__(self, name):
        super().__init__(name, 600)  # è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        self.ammo = "å¤§å¼¹ä¸¸"

    def attack(self, target: StandardRobot):
        """é‡å†™æ”»å‡»æ–¹æ³•"""
        damage = 10  # è‹±é›„ä¼¤å®³æ›´é«˜
        target.hp -= damage
        print(f"{self.name} ä½¿ç”¨{self.ammo}æ”»å‡» {target.name}ï¼Œé€ æˆ {damage} ç‚¹ä¼¤å®³")


class InfantryRobot(StandardRobot):
    """å“¨å…µæœºå™¨äººç±»ï¼Œç»§æ‰¿è‡ªStandardRobot"""

    def __init__(self, name):
        super().__init__(name, 200)  # è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        self.ammo = "å°å¼¹ä¸¸"

    def rapid_fire(self, target: StandardRobot, shots=3):
        """å®šä¹‰å­ç±»ç‰¹æœ‰æ–¹æ³• - è¿å‘"""
        total_damage = 0
        for i in range(shots):
            damage = 3
            target.hp -= damage
            total_damage += damage
        print(f"{self.name} è¿å‘ {shots} æ¬¡ï¼Œæ€»å…±é€ æˆ {total_damage} ç‚¹ä¼¤å®³")


# æµ‹è¯•ç»§æ‰¿
# åˆ›å»ºä¸åŒç±»å‹çš„æœºå™¨äºº
hero = HeroRobot("Hero-Alpha")
infantry = InfantryRobot("Infantry-Beta")
standard = StandardRobot("Standard-Robot", 150)

print(f"è‹±é›„æœºå™¨äºº: {hero.name}, è¡€é‡: {hero.hp}, å¼¹è¯: {hero.ammo}")
print(f"å“¨å…µæœºå™¨äºº: {infantry.name}, è¡€é‡: {infantry.hp}, å¼¹è¯: {infantry.ammo}")
print(f"æ­¥å…µæœºå™¨äºº: {standard.name}, è¡€é‡: {standard.hp}")

# æµ‹è¯•ç»§æ‰¿çš„æ–¹æ³•
print(f"\n=== æ–¹æ³•è°ƒç”¨æµ‹è¯• ===")
hero.move(5, 5)           # ç»§æ‰¿çš„æ–¹æ³•
hero.attack(infantry)     # é‡å†™çš„æ–¹æ³•

infantry.rapid_fire(hero)  # å­ç±»ç‰¹æœ‰æ–¹æ³•

standard.attack(hero)   # åŸºç±»æ–¹æ³•

print(f"\næˆ˜æ–—åè¡€é‡:")
print(f"{hero.name}: {hero.hp}HP")
print(f"{infantry.name}: {infantry.hp}HP")
```

### 2.3 å¤šæ€ (Polymorphism)

å¤šæ€æ˜¯æŒ‡åŒä¸€ä¸ªæ–¹æ³•åœ¨ä¸åŒç±»ä¸­å¯ä»¥æœ‰ä¸åŒçš„å®ç°ï¼Œè°ƒç”¨æ—¶ä¼šæ ¹æ®å¯¹è±¡çš„å®é™…ç±»å‹æ‰§è¡Œç›¸åº”çš„æ–¹æ³•ã€‚

```python
# å¤šæ€ - åŒä¸€æ¥å£ï¼Œä¸åŒå®ç°

class StandardRobot:
    """æœºå™¨äººåŸºç±»"""

    def __init__(self, name, hp):
        self.name = name
        self.hp = hp

    def attack(self, target):
        """æ”»å‡»æ–¹æ³• - å­ç±»åº”è¯¥é‡å†™è¿™ä¸ªæ–¹æ³•"""
        print(f"{self.name} è¿›è¡ŒåŸºç¡€æ”»å‡»")

    def special_ability(self):
        """ç‰¹æ®Šèƒ½åŠ› - å­ç±»åº”è¯¥é‡å†™è¿™ä¸ªæ–¹æ³•"""
        print(f"{self.name} ä½¿ç”¨åŸºç¡€èƒ½åŠ›")

# ä¸åŒçš„å­ç±»å®ç°ä¸åŒçš„æ”»å‡»æ–¹å¼


class HeroRobot(StandardRobot):
    def attack(self, target):
        print(f"ğŸ”¥ {self.name} å‘å°„å¤§å¼¹ä¸¸æ”»å‡» {target.name}ï¼")

    def special_ability(self):
        print(f"âš¡ {self.name} ä½¿ç”¨æŠ¤ç›¾æŠ€èƒ½ï¼")


class InfantryRobot(StandardRobot):
    def attack(self, target):
        print(f"ğŸ”« {self.name} è¿ç»­å°„å‡»æ”»å‡» {target.name}ï¼")

    def special_ability(self):
        print(f"ğŸ’¨ {self.name} ä½¿ç”¨åŠ é€Ÿå†²åˆºï¼")


class EngineerRobot(StandardRobot):
    def attack(self, target):
        print(f"ğŸ”§ {self.name} æ— æ³•æ”»å‡»ï¼Œä½†å¯ä»¥ä¿®å¤ç›Ÿå‹ï¼")

    def special_ability(self):
        print(f"â›ï¸ {self.name} å¼€å§‹å…‘çŸ¿æ“ä½œï¼")


# åˆ›å»ºä¸åŒç±»å‹çš„æœºå™¨äºº
robots = [
    HeroRobot("Hero-1", 600),
    InfantryRobot("Infantry-1", 200),
    EngineerRobot("Engineer-1", 250)
]

# ä½¿ç”¨ç›¸åŒçš„æ¥å£è°ƒç”¨ï¼Œä½†è¡¨ç°å‡ºä¸åŒçš„è¡Œä¸º
target = StandardRobot("Target", 100)

for robot in robots:
    print(f"\n{robot.name} çš„è¡Œä¸º:")
    robot.attack(target)        # å¤šæ€ï¼šç›¸åŒæ–¹æ³•åï¼Œä¸åŒå®ç°
    robot.special_ability()     # å¤šæ€ï¼šç›¸åŒæ–¹æ³•åï¼Œä¸åŒå®ç°

# å¤šæ€çš„å®é™…åº”ç”¨ - ç»Ÿä¸€å¤„ç†ä¸åŒç±»å‹çš„å¯¹è±¡


def command_robot_attack(robot, target):
    """ç»Ÿä¸€çš„æœºå™¨äººæ”»å‡»æŒ‡ä»¤ - å¤šæ€çš„ä½“ç°"""
    print(f"æŒ‡æŒ¥å®˜: {robot.name}ï¼Œæ”»å‡»ç›®æ ‡ï¼")
    robot.attack(target)  # ä¸éœ€è¦çŸ¥é“å…·ä½“æ˜¯ä»€ä¹ˆç±»å‹çš„æœºå™¨äºº


def command_special_ability(robot):
    """ç»Ÿä¸€çš„ç‰¹æ®Šèƒ½åŠ›æŒ‡ä»¤"""
    print(f"æŒ‡æŒ¥å®˜: {robot.name}ï¼Œä½¿ç”¨ç‰¹æ®Šèƒ½åŠ›ï¼")
    robot.special_ability()  # ä¸éœ€è¦çŸ¥é“å…·ä½“æ˜¯ä»€ä¹ˆç±»å‹çš„æœºå™¨äºº


print(f"\n=== å¤šæ€çš„å®é™…åº”ç”¨ï¼šç»Ÿä¸€æŒ‡æŒ¥æ¥å£ ===")
for robot in robots:
    command_robot_attack(robot, target)
    command_special_ability(robot)
    print()

print("âœ… å¤šæ€ç‰¹æ€§æ¼”ç¤ºå®Œæˆï¼")
```

```python
# å¼‚å¸¸å¤„ç†åŸºç¡€ - try, except, finally, raise

print("=== å¼‚å¸¸å¤„ç†åŸºç¡€è¯­æ³• ===")

# 1. åŸºæœ¬çš„ try-except ç»“æ„
print("\n--- åŸºæœ¬å¼‚å¸¸æ•è· ---")

def divide_numbers(a, b):
    """é™¤æ³•è¿ç®—ç¤ºä¾‹"""
    try:
        result = a / b
        print(f"{a} Ã· {b} = {result}")
        return result
    except ZeroDivisionError:
        print("âŒ é”™è¯¯ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼")
        return None
    except TypeError:
        print("âŒ é”™è¯¯ï¼šå‚æ•°ç±»å‹é”™è¯¯ï¼")
        return None

# æµ‹è¯•åŸºæœ¬å¼‚å¸¸å¤„ç†
divide_numbers(10, 2)    # æ­£å¸¸æƒ…å†µ
divide_numbers(10, 0)    # é™¤é›¶å¼‚å¸¸
divide_numbers(10, "a")  # ç±»å‹å¼‚å¸¸
```

```python
# 2. æ•è·å¤šç§å¼‚å¸¸
print("\n--- æ•è·å¤šç§å¼‚å¸¸ ---")


def process_robot_data(robot_data):
    """å¤„ç†æœºå™¨äººæ•°æ®"""
    try:
        name = robot_data["name"]
        hp = int(robot_data["hp"])
        damage = float(robot_data["damage"])
        print(f"æœºå™¨äºº {name}: HP={hp}, æ”»å‡»åŠ›={damage}")
        return True
    except KeyError as e:
        print(f"âŒ ç¼ºå°‘å¿…è¦å­—æ®µ: {e}")
    except ValueError as e:
        print(f"âŒ æ•°æ®ç±»å‹é”™è¯¯: {e}")
    except Exception as e:
        print(f"âŒ æœªçŸ¥é”™è¯¯: {e}")
    return False


# æµ‹è¯•å¤šç§å¼‚å¸¸
good_data = {"name": "Hero-1", "hp": "600", "damage": "10.5"}
bad_data1 = {"name": "Hero-2", "damage": "10"}  # ç¼ºå°‘hpå­—æ®µ
bad_data2 = {"name": "Hero-3", "hp": "abc", "damage": "10"}  # hpä¸æ˜¯æ•°å­—

process_robot_data(good_data)
process_robot_data(bad_data1)
process_robot_data(bad_data2)
```

```python
# 3. finally å­å¥ - æ— è®ºæ˜¯å¦å‡ºç°å¼‚å¸¸éƒ½ä¼šæ‰§è¡Œ
print("\n--- finally å­å¥ ---")


def save_robot_state(robot_name, state):
    """ä¿å­˜æœºå™¨äººçŠ¶æ€ï¼ˆæ¨¡æ‹Ÿï¼‰"""
    file_handle = None
    try:
        print(f"å¼€å§‹ä¿å­˜ {robot_name} çš„çŠ¶æ€...")
        file_handle = f"æ¨¡æ‹Ÿæ–‡ä»¶å¥æŸ„_{robot_name}"

        if state is None:
            raise ValueError("çŠ¶æ€æ•°æ®ä¸èƒ½ä¸ºç©º")

        print(f"âœ… {robot_name} çŠ¶æ€ä¿å­˜æˆåŠŸ")
        return True

    except ValueError as e:
        print(f"âŒ ä¿å­˜å¤±è´¥: {e}")
        return False
    finally:
        # æ— è®ºæˆåŠŸå¤±è´¥éƒ½è¦æ‰§è¡Œæ¸…ç†å·¥ä½œ
        if file_handle:
            print(f"ğŸ”§ æ¸…ç†èµ„æºï¼šå…³é—­æ–‡ä»¶å¥æŸ„ {file_handle}")


# æµ‹è¯• finally
save_robot_state("Hero-1", {"hp": 100})  # æ­£å¸¸æƒ…å†µ
save_robot_state("Hero-2", None)         # å¼‚å¸¸æƒ…å†µ
```

```python
# 4. ä¸»åŠ¨æŠ›å‡ºå¼‚å¸¸ - raise
print("\n--- ä¸»åŠ¨æŠ›å‡ºå¼‚å¸¸ ---")


def validate_robot_hp(hp):
    """éªŒè¯æœºå™¨äººè¡€é‡"""
    if not isinstance(hp, (int, float)):
        raise TypeError("è¡€é‡å¿…é¡»æ˜¯æ•°å­—ç±»å‹")

    if hp < 0:
        raise ValueError("è¡€é‡ä¸èƒ½ä¸ºè´Ÿæ•°")

    if hp > 1000:
        raise ValueError("è¡€é‡ä¸èƒ½è¶…è¿‡1000")

    print(f"âœ… è¡€é‡éªŒè¯é€šè¿‡: {hp}")


# æµ‹è¯•ä¸»åŠ¨æŠ›å‡ºå¼‚å¸¸
try:
    validate_robot_hp(600)      # æ­£å¸¸
    validate_robot_hp(-50)      # è´Ÿæ•°
except ValueError as e:
    print(f"âŒ è¡€é‡éªŒè¯å¤±è´¥: {e}")

try:
    validate_robot_hp("abc")    # ç±»å‹é”™è¯¯
except TypeError as e:
    print(f"âŒ ç±»å‹éªŒè¯å¤±è´¥: {e}")
```

```python
# è‡ªå®šä¹‰å¼‚å¸¸ç±»

print("=== è‡ªå®šä¹‰å¼‚å¸¸ç±» ===")

# 1. å®šä¹‰è‡ªå®šä¹‰å¼‚å¸¸
class RobotError(Exception):
    """æœºå™¨äººç›¸å…³å¼‚å¸¸çš„åŸºç±»"""
    def __init__(self, message, robot_name=None):
        super().__init__(message)
        self.robot_name = robot_name
    
    def __str__(self):
        if self.robot_name:
            return f"æœºå™¨äºº {self.robot_name}: {super().__str__()}"
        return super().__str__()

class RobotDestroyedException(RobotError):
    """æœºå™¨äººå·²æŸåå¼‚å¸¸"""
    pass

class InsufficientEnergyException(RobotError):
    """èƒ½é‡ä¸è¶³å¼‚å¸¸"""
    def __init__(self, message, robot_name=None, current_energy=0, required_energy=0):
        super().__init__(message, robot_name)
        self.current_energy = current_energy
        self.required_energy = required_energy

class InvalidPositionException(RobotError):
    """æ— æ•ˆä½ç½®å¼‚å¸¸"""
    pass

# 2. ä½¿ç”¨è‡ªå®šä¹‰å¼‚å¸¸
def check_robot_energy(robot_name, current_energy, required_energy):
    """æ£€æŸ¥æœºå™¨äººèƒ½é‡"""
    try:
        if current_energy < required_energy:
            raise InsufficientEnergyException(
                f"èƒ½é‡ä¸è¶³ï¼šéœ€è¦ {required_energy}ï¼Œå½“å‰ {current_energy}",
                robot_name, current_energy, required_energy
            )
        print(f"âœ… {robot_name} èƒ½é‡å……è¶³")
        return True
    except InsufficientEnergyException as e:
        print(f"âŒ {e}")
        print(f"   è¯¦ç»†ä¿¡æ¯ï¼šå½“å‰={e.current_energy}, éœ€è¦={e.required_energy}")
        return False

def move_robot(robot_name, position, field_bounds=(100, 100)):
    """ç§»åŠ¨æœºå™¨äºº"""
    x, y = position
    max_x, max_y = field_bounds
    
    try:
        if not (0 <= x <= max_x and 0 <= y <= max_y):
            raise InvalidPositionException(
                f"ä½ç½® ({x}, {y}) è¶…å‡ºåœºåœ°èŒƒå›´ (0-{max_x}, 0-{max_y})",
                robot_name
            )
        print(f"âœ… {robot_name} ç§»åŠ¨åˆ°ä½ç½® ({x}, {y})")
        return True
    except InvalidPositionException as e:
        print(f"âŒ {e}")
        return False

# æµ‹è¯•è‡ªå®šä¹‰å¼‚å¸¸
print("\n--- æµ‹è¯•è‡ªå®šä¹‰å¼‚å¸¸ ---")

# èƒ½é‡æ£€æŸ¥æµ‹è¯•
check_robot_energy("Hero-1", 80, 50)    # å……è¶³
check_robot_energy("Hero-2", 20, 50)    # ä¸è¶³

# ä½ç½®æ£€æŸ¥æµ‹è¯•
move_robot("Infantry-1", (50, 30))      # æ­£å¸¸ä½ç½®
move_robot("Infantry-2", (150, 200))    # è¶…å‡ºèŒƒå›´
```

```python
# åœ¨ç±»ä¸­æ•´åˆå¼‚å¸¸å¤„ç†

print("=== åœ¨ç±»ä¸­æ•´åˆå¼‚å¸¸å¤„ç† ===")

# å¸¦å¼‚å¸¸å¤„ç†çš„æœºå™¨äººç±»
class SafeRobot:
    """é›†æˆå¼‚å¸¸å¤„ç†çš„å®‰å…¨æœºå™¨äººç±»"""
    
    def __init__(self, name, robot_type):
        self.name = name
        self.robot_type = robot_type
        self.hp = 100
        self.max_hp = 100
        self.energy = 100
        self.position = (0, 0)
        self.is_destroyed = False
    
    def _check_if_destroyed(self):
        """æ£€æŸ¥æœºå™¨äººæ˜¯å¦å·²æŸå"""
        if self.is_destroyed:
            raise RobotDestroyedException(f"æœºå™¨äººå·²æŸåï¼Œæ— æ³•æ‰§è¡Œæ“ä½œ", self.name)
    
    def move(self, x, y, field_bounds=(100, 100)):
        """å®‰å…¨ç§»åŠ¨æ–¹æ³•"""
        try:
            self._check_if_destroyed()
            
            # æ£€æŸ¥ä½ç½®æœ‰æ•ˆæ€§
            max_x, max_y = field_bounds
            if not (0 <= x <= max_x and 0 <= y <= max_y):
                raise InvalidPositionException(
                    f"ç›®æ ‡ä½ç½® ({x}, {y}) è¶…å‡ºåœºåœ°èŒƒå›´",
                    self.name
                )
            
            # æ£€æŸ¥èƒ½é‡
            energy_cost = 10
            if self.energy < energy_cost:
                raise InsufficientEnergyException(
                    "ç§»åŠ¨èƒ½é‡ä¸è¶³",
                    self.name, self.energy, energy_cost
                )
            
            # æ‰§è¡Œç§»åŠ¨
            old_pos = self.position
            self.position = (x, y)
            self.energy -= energy_cost
            
            print(f"âœ… {self.name} ä» {old_pos} ç§»åŠ¨åˆ° ({x}, {y})ï¼Œæ¶ˆè€— {energy_cost} èƒ½é‡")
            return True
            
        except (RobotDestroyedException, InvalidPositionException, InsufficientEnergyException) as e:
            print(f"âŒ ç§»åŠ¨å¤±è´¥: {e}")
            return False
    
    def attack(self, target):
        """å®‰å…¨æ”»å‡»æ–¹æ³•"""
        try:
            self._check_if_destroyed()
            
            if not isinstance(target, SafeRobot):
                raise TypeError("æ”»å‡»ç›®æ ‡å¿…é¡»æ˜¯SafeRobotç±»å‹")
            
            if target.is_destroyed:
                raise RobotDestroyedException("ç›®æ ‡å·²æŸå", target.name)
            
            # æ£€æŸ¥èƒ½é‡
            energy_cost = 20
            if self.energy < energy_cost:
                raise InsufficientEnergyException(
                    "æ”»å‡»èƒ½é‡ä¸è¶³",
                    self.name, self.energy, energy_cost
                )
            
            # æ‰§è¡Œæ”»å‡»
            damage = 30
            self.energy -= energy_cost
            target.take_damage(damage)
            
            print(f"ğŸ”¥ {self.name} æ”»å‡» {target.name}ï¼Œé€ æˆ {damage} ç‚¹ä¼¤å®³")
            return True
            
        except (RobotDestroyedException, InsufficientEnergyException, TypeError) as e:
            print(f"âŒ æ”»å‡»å¤±è´¥: {e}")
            return False
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        try:
            if damage < 0:
                raise ValueError("ä¼¤å®³å€¼ä¸èƒ½ä¸ºè´Ÿæ•°")
            
            old_hp = self.hp
            self.hp = max(0, self.hp - damage)
            
            print(f"ğŸ’¥ {self.name} å—åˆ° {damage} ä¼¤å®³: {old_hp} -> {self.hp}")
            
            if self.hp == 0:
                self.is_destroyed = True
                print(f"ğŸ’€ {self.name} è¢«æ‘§æ¯ï¼")
            
        except ValueError as e:
            print(f"âŒ ä¼¤å®³å¤„ç†é”™è¯¯: {e}")
    
    def repair(self, heal_amount):
        """ä¿®å¤æœºå™¨äºº"""
        try:
            if self.is_destroyed:
                raise RobotDestroyedException("æœºå™¨äººå·²å®Œå…¨æŸåï¼Œæ— æ³•ä¿®å¤", self.name)
            
            if heal_amount <= 0:
                raise ValueError("ä¿®å¤é‡å¿…é¡»å¤§äº0")
            
            old_hp = self.hp
            self.hp = min(self.max_hp, self.hp + heal_amount)
            actual_heal = self.hp - old_hp
            
            print(f"ğŸ”§ {self.name} ä¿®å¤ {actual_heal} ç‚¹è¡€é‡: {old_hp} -> {self.hp}")
            return actual_heal
            
        except (RobotDestroyedException, ValueError) as e:
            print(f"âŒ ä¿®å¤å¤±è´¥: {e}")
            return 0

# æµ‹è¯•é›†æˆå¼‚å¸¸å¤„ç†çš„æœºå™¨äººç±»
print("\n=== æµ‹è¯•é›†æˆå¼‚å¸¸å¤„ç† ===")

# åˆ›å»ºæœºå™¨äºº
robot1 = SafeRobot("Hero-Alpha", "Hero")
robot2 = SafeRobot("Infantry-Beta", "Infantry")

print("\n--- æ­£å¸¸æ“ä½œæµ‹è¯• ---")
robot1.move(10, 20)
robot1.attack(robot2)
robot2.repair(10)

print("\n--- å¼‚å¸¸æƒ…å†µæµ‹è¯• ---")
# æ¶ˆè€—èƒ½é‡ç›´åˆ°ä¸è¶³
for i in range(8):  # æ¶ˆè€—å¤§é‡èƒ½é‡
    robot1.move(i, i)

# èƒ½é‡ä¸è¶³åå°è¯•æ”»å‡»
robot1.attack(robot2)

# å°†robot2æ‰“è‡³æŸå
robot2.take_damage(200)  # å¤§é‡ä¼¤å®³

# å°è¯•å¯¹å·²æŸåçš„æœºå™¨äººè¿›è¡Œæ“ä½œ
robot2.move(30, 40)      # å·²æŸåï¼Œæ— æ³•ç§»åŠ¨
robot1.attack(robot2)    # æ”»å‡»å·²æŸåçš„ç›®æ ‡
robot2.repair(50)        # å°è¯•ä¿®å¤å·²æŸåçš„æœºå™¨äºº
```

## ç¬¬å››éƒ¨åˆ†ï¼šç»¼åˆå®è·µé¡¹ç›® - æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨

è®©æˆ‘ä»¬åˆ›å»ºä¸€ä¸ªå®Œæ•´çš„æœºå™¨äººæ§åˆ¶ç³»ç»Ÿï¼Œç»¼åˆè¿ç”¨æ‰€æœ‰å­¦åˆ°çš„çŸ¥è¯†ã€‚

```python
# ç»¼åˆå®è·µé¡¹ç›®ï¼šç®€åŒ–ç‰ˆæœºå™¨äººç®¡ç†ç³»ç»Ÿ

print("ğŸ¤– æœºå™¨äººç®¡ç†ç³»ç»Ÿ")
print("=" * 50)

# ç»¼åˆè¿ç”¨æ‰€æœ‰å­¦åˆ°çš„çŸ¥è¯†ç‚¹
class RobotManager:
    """æœºå™¨äººç®¡ç†ç³»ç»Ÿ"""
    
    def __init__(self):
        self.robots = {}
        self.battle_log = []
    
    def create_robot(self, robot_id, name, robot_type):
        """åˆ›å»ºæœºå™¨äºº"""
        try:
            if robot_id in self.robots:
                raise ValueError(f"æœºå™¨äººID {robot_id} å·²å­˜åœ¨")
            
            # æ ¹æ®ç±»å‹åˆ›å»ºä¸åŒçš„æœºå™¨äºº
            if robot_type == "Hero":
                robot = HeroRobot(name)
            elif robot_type == "Infantry":
                robot = InfantryRobot(name)
            elif robot_type == "Engineer":
                robot = EngineerRobot(name)
            else:
                raise ValueError(f"æœªçŸ¥æœºå™¨äººç±»å‹: {robot_type}")
            
            self.robots[robot_id] = robot
            print(f"âœ… åˆ›å»ºæˆåŠŸ: {name} ({robot_type})")
            return True
            
        except ValueError as e:
            print(f"âŒ åˆ›å»ºå¤±è´¥: {e}")
            return False
    
    def battle_simulation(self, attacker_id, target_id):
        """æˆ˜æ–—æ¨¡æ‹Ÿ"""
        try:
            if attacker_id not in self.robots or target_id not in self.robots:
                raise KeyError("æœºå™¨äººä¸å­˜åœ¨")
            
            attacker = self.robots[attacker_id]
            target = self.robots[target_id]
            
            # æ‰§è¡Œæ”»å‡»
            success = attacker.attack(target)
            if success:
                log_entry = f"{attacker.name} æ”»å‡»äº† {target.name}"
                self.battle_log.append(log_entry)
            
            return success
            
        except Exception as e:
            print(f"âŒ æˆ˜æ–—å¤±è´¥: {e}")
            return False
    
    def display_status(self):
        """æ˜¾ç¤ºæ‰€æœ‰æœºå™¨äººçŠ¶æ€"""
        print(f"\nğŸ“Š æœºå™¨äººçŠ¶æ€æŠ¥å‘Š:")
        print("-" * 40)
        for robot_id, robot in self.robots.items():
            status = "æ­£å¸¸" if robot.hp > 0 else "æŸå"
            extra_info = ""
            if hasattr(robot, 'ore_count'):
                extra_info = f", çŸ¿çŸ³: {robot.ore_count}"
            print(f"  {robot_id}: {robot.name} - HP: {robot.hp}/{robot.max_hp}, çŠ¶æ€: {status}{extra_info}")
    
    def get_battle_log(self):
        """è·å–æˆ˜æ–—æ—¥å¿—"""
        print(f"\nğŸ“œ æˆ˜æ–—æ—¥å¿—:")
        for i, log in enumerate(self.battle_log, 1):
            print(f"  {i}. {log}")

# ç®€åŒ–çš„æœºå™¨äººç±»
class SimpleRobot:
    """ç®€åŒ–çš„æœºå™¨äººåŸºç±»"""
    
    def __init__(self, name, hp, damage):
        self.name = name
        self.hp = hp
        self.max_hp = hp
        self.damage = damage
    
    def attack(self, target):
        """æ”»å‡»æ–¹æ³•"""
        if self.hp <= 0:
            print(f"âŒ {self.name} å·²æŸåï¼Œæ— æ³•æ”»å‡»")
            return False
        
        target.hp = max(0, target.hp - self.damage)
        print(f"ğŸ”¥ {self.name} æ”»å‡» {target.name}ï¼Œé€ æˆ {self.damage} ç‚¹ä¼¤å®³")
        
        if target.hp == 0:
            print(f"ğŸ’€ {target.name} è¢«å‡»è´¥ï¼")
        
        return True

class HeroRobot(SimpleRobot):
    """è‹±é›„æœºå™¨äºº"""
    
    def __init__(self, name):
        super().__init__(name, 600, 10)  # é«˜è¡€é‡ï¼Œé«˜æ”»å‡»
    
    def special_attack(self, target):
        """ç‰¹æ®Šæ”»å‡»"""
        if self.hp <= 0:
            return False
        
        damage = self.damage * 2  # åŒå€ä¼¤å®³
        target.hp = max(0, target.hp - damage)
        print(f"âš¡ {self.name} ä½¿ç”¨ç‰¹æ®Šæ”»å‡»å¯¹ {target.name} é€ æˆ {damage} ç‚¹ä¼¤å®³ï¼")
        return True

class InfantryRobot(SimpleRobot):
    """å“¨å…µæœºå™¨äºº"""
    
    def __init__(self, name):
        super().__init__(name, 200, 3)  # æ ‡å‡†è¡€é‡ï¼Œä¸­ç­‰æ”»å‡»

class EngineerRobot(SimpleRobot):
    """å·¥ç¨‹æœºå™¨äºº"""
    
    def __init__(self, name):
        super().__init__(name, 250, 0)  # ä¸­ç­‰è¡€é‡ï¼Œæ— æ”»å‡»
        self.ore_count = 0
    
    def attack(self, target):
        """å·¥ç¨‹æœºå™¨äººæ— æ³•æ”»å‡»"""
        print(f"âŒ {self.name} æ˜¯å·¥ç¨‹æœºå™¨äººï¼Œæ— æ³•æ”»å‡»")
        return False
    
    def mine_ore(self):
        """å…‘çŸ¿"""
        if self.hp <= 0:
            print(f"âŒ {self.name} å·²æŸåï¼Œæ— æ³•å…‘çŸ¿")
            return False
        
        ore_amount = 20
        self.ore_count += ore_amount
        print(f"â›ï¸ {self.name} å…‘å–äº† {ore_amount} çŸ¿çŸ³ï¼Œæ€»è®¡: {self.ore_count}")
        return True
    
    def repair_ally(self, target):
        """ä¿®å¤ç›Ÿå‹"""
        if self.hp <= 0 or self.ore_count < 10:
            print(f"âŒ {self.name} æ— æ³•ä¿®å¤ï¼ˆæŸåæˆ–çŸ¿çŸ³ä¸è¶³ï¼‰")
            return False
        
        heal_amount = 50
        old_hp = target.hp
        target.hp = min(target.max_hp, target.hp + heal_amount)
        actual_heal = target.hp - old_hp
        self.ore_count -= 10  # æ¶ˆè€—çŸ¿çŸ³
        
        print(f"ğŸ”§ {self.name} ä¿®å¤ {target.name}: +{actual_heal}HPï¼ˆæ¶ˆè€—10çŸ¿çŸ³ï¼‰")
        return True

# è¿è¡Œæ¼”ç¤º
print("\n=== åˆ›å»ºæœºå™¨äººé˜Ÿä¼ ===")
manager = RobotManager()

# åˆ›å»ºä¸åŒç±»å‹çš„æœºå™¨äºº
manager.create_robot("H01", "Hero-Alpha", "Hero")
manager.create_robot("I01", "Infantry-Beta", "Infantry")
manager.create_robot("I02", "Infantry-Gamma", "Infantry")
manager.create_robot("E01", "Engineer-Delta", "Engineer")

# æ˜¾ç¤ºåˆå§‹çŠ¶æ€
manager.display_status()

# æ¼”ç¤ºå„ç§æ“ä½œ
print(f"\n=== æ“ä½œæ¼”ç¤º ===")

# è‹±é›„ç‰¹æ®Šæ”»å‡»
hero = manager.robots["H01"]
hero.special_attack(manager.robots["I01"])

# æ™®é€šæ”»å‡»
manager.battle_simulation("I02", "H01")

# å·¥ç¨‹æœºå™¨äººå…‘çŸ¿
engineer = manager.robots["E01"]
engineer.mine_ore()
engineer.mine_ore()

# ä¿®å¤ç›Ÿå‹
engineer.repair_ally(manager.robots["I01"])

# æ˜¾ç¤ºæœ€ç»ˆçŠ¶æ€
manager.display_status()
manager.get_battle_log()
```
