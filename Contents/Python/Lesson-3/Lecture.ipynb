### 1.1 ä»€ä¹ˆæ˜¯é¢å‘å¯¹è±¡ç¼–ç¨‹ (OOP)

é¢å‘å¯¹è±¡ç¼–ç¨‹æ˜¯ä¸€ç§ç¼–ç¨‹èŒƒå¼ï¼Œå®ƒå°†ç°å®ä¸–ç•Œçš„äº‹ç‰©æŠ½è±¡ä¸ºç¨‹åºä¸­çš„å¯¹è±¡ã€‚

**æ ¸å¿ƒæ¦‚å¿µï¼š**
- **å¯¹è±¡ (Object)**: ç°å®ä¸–ç•Œå®ä½“çš„æŠ½è±¡ï¼ŒåŒ…å«æ•°æ®å’Œæ–¹æ³•
- **ç±» (Class)**: å¯¹è±¡çš„æ¨¡æ¿æˆ–è“å›¾ï¼Œå®šä¹‰å¯¹è±¡çš„ç»“æ„
- **å®ä¾‹ (Instance)**: ç±»çš„å…·ä½“å®ç°ï¼Œä¸€ä¸ªçœŸå®å­˜åœ¨çš„å¯¹è±¡

**åœ¨ RoboMaster ä¸­çš„å¯¹è±¡æ€ç»´ï¼š**
- æœºå™¨äºº â†’ å¯¹è±¡
- æœºå™¨äººç±»å‹ï¼ˆHeroã€Infantryç­‰ï¼‰ â†’ ç±»

```python
# ç¬¬ä¸€ä¸ªå®Œæ•´çš„ç±»å®šä¹‰ - åŸºäºçœŸå®RoboMasterè§„åˆ™

print("=== é¢å‘å¯¹è±¡ç¼–ç¨‹åŸºç¡€ ===")

class Robot:
    """æœºå™¨äººåŸºç±» - åŸºäºçœŸå®RoboMasterè§„åˆ™è®¾è®¡"""
    
    # ç±»å±æ€§ï¼ˆæ‰€æœ‰å®ä¾‹å…±äº«ï¼‰
    game_version = "2024"
    robot_count = 0
    
    def __init__(self, name, robot_type, hp):
        """
        æ„é€ å‡½æ•°ï¼šåˆ›å»ºæœºå™¨äººå®ä¾‹æ—¶è‡ªåŠ¨è°ƒç”¨
        
        Args:
            name (str): æœºå™¨äººåç§°
            robot_type (str): æœºå™¨äººç±»å‹ (Hero/Infantry/Engineer)
            hp (int): ç”Ÿå‘½å€¼
        """
        # å®ä¾‹å±æ€§ï¼ˆæ¯ä¸ªå®ä¾‹ç‹¬æœ‰ï¼‰
        self.name = name
        self.robot_type = robot_type
        self.hp = hp
        self.max_hp = hp
        self.position = (0, 0)
        self.status = "å¾…å‘½"
        
        # æ ¹æ®æœºå™¨äººç±»å‹è®¾ç½®æ”»å‡»åŠ›
        if robot_type == "Hero":
            self.damage = 10  # è‹±é›„æœºå™¨äººå¤§å¼¹ä¸¸ï¼Œ10ç‚¹ä¼¤å®³
        elif robot_type == "Infantry":
            self.damage = 1   # æ­¥å…µæœºå™¨äººå°å¼¹ä¸¸ï¼Œ1ç‚¹ä¼¤å®³
        else:  # Engineer
            self.damage = 0   # å·¥ç¨‹æœºå™¨äººæ— æ”»å‡»èƒ½åŠ›
        
        # æ›´æ–°ç±»å±æ€§
        Robot.robot_count += 1
        
        print(f"âœ… åˆ›å»º{robot_type}æœºå™¨äºº: {name} (è¡€é‡:{hp}HP, æ”»å‡»:{self.damage}ç‚¹)")
    
    def __str__(self):
        """å­—ç¬¦ä¸²è¡¨ç¤º - ç”¨äºprint()"""
        return f"{self.robot_type}({self.name}, HP:{self.hp}/{self.max_hp}, æ”»å‡»:{self.damage}ç‚¹)"
    
    def __repr__(self):
        """å¼€å‘è€…è¡¨ç¤º - ç”¨äºè°ƒè¯•"""
        return f"Robot('{self.name}', '{self.robot_type}', {self.hp})"
    
    # å®ä¾‹æ–¹æ³•
    def move(self, x, y):
        """ç§»åŠ¨æœºå™¨äººåˆ°æŒ‡å®šä½ç½®"""
        if self.status == "å¾…å‘½":
            self.position = (x, y)
            print(f"{self.name} ç§»åŠ¨åˆ°ä½ç½® ({x}, {y})")
        else:
            print(f"âŒ {self.name} çŠ¶æ€å¼‚å¸¸ï¼Œæ— æ³•ç§»åŠ¨")
    
    def attack(self, target):
        """æ”»å‡»ç›®æ ‡æœºå™¨äºº"""
        if self.damage == 0:
            print(f"âŒ {self.name} æ˜¯å·¥ç¨‹æœºå™¨äººï¼Œæ— æ³•æ”»å‡»")
            return 0
        
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} çŠ¶æ€å¼‚å¸¸ï¼Œæ— æ³•æ”»å‡»")
            return 0
        
        print(f"ğŸ”¥ {self.name} æ”»å‡» {target.name}ï¼Œé€ æˆ {self.damage} ç‚¹ä¼¤å®³")
        target.take_damage(self.damage)
        return self.damage
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        old_hp = self.hp
        self.hp = max(0, self.hp - damage)
        print(f"ğŸ’¥ {self.name} å—åˆ° {damage} ç‚¹ä¼¤å®³: {old_hp} -> {self.hp}")
        
        if self.hp == 0:
            self.status = "æŸå"
            print(f"� {self.name} å·²è¢«æ‘§æ¯!")
    
    def repair(self, heal_amount):
        """ä¿®å¤æœºå™¨äºº"""
        if self.status == "æŸå":
            print(f"âŒ {self.name} å·²æŸåï¼Œéœ€è¦å®Œå…¨é‡å»º")
            return
        
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + heal_amount)
        print(f"ğŸ”§ {self.name} ä¿®å¤ {heal_amount} ç‚¹: {old_hp} -> {self.hp}")
    
    def mine_ore(self):
        """å…‘çŸ¿æ“ä½œï¼ˆä»…å·¥ç¨‹æœºå™¨äººå¯ç”¨ï¼‰"""
        if self.robot_type != "Engineer":
            print(f"âŒ {self.name} ä¸æ˜¯å·¥ç¨‹æœºå™¨äººï¼Œæ— æ³•å…‘çŸ¿")
            return False
        
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} çŠ¶æ€å¼‚å¸¸ï¼Œæ— æ³•å…‘çŸ¿")
            return False
        
        print(f"â›ï¸ {self.name} æ‰§è¡Œå…‘çŸ¿æ“ä½œ")
        return True
    
    def get_info(self):
        """è·å–æœºå™¨äººè¯¦ç»†ä¿¡æ¯"""
        return {
            "name": self.name,
            "type": self.robot_type,
            "hp": f"{self.hp}/{self.max_hp}",
            "damage": self.damage,
            "position": self.position,
            "status": self.status
        }

# åˆ›å»ºä¸åŒç±»å‹çš„æœºå™¨äºº
print("\n=== åˆ›å»ºæœºå™¨äººé˜Ÿä¼ ===")
hero_robot = Robot("Hero-1", "Hero", 600)      # è‹±é›„æœºå™¨äººï¼šè¡€é‡åšï¼Œæ”»å‡»å¼º
infantry_1 = Robot("Infantry-1", "Infantry", 200)  # æ­¥å…µæœºå™¨äººï¼šæ ‡å‡†è¡€é‡ï¼Œæ”»å‡»å¼±
infantry_2 = Robot("Infantry-2", "Infantry", 200)  # ç¬¬äºŒä¸ªæ­¥å…µ
engineer = Robot("Engineer-1", "Engineer", 250)    # å·¥ç¨‹æœºå™¨äººï¼šä¸­ç­‰è¡€é‡ï¼Œæ— æ”»å‡»

# æ˜¾ç¤ºæ‰€æœ‰æœºå™¨äººä¿¡æ¯
print(f"\n=== æœºå™¨äººä¿¡æ¯æ€»è§ˆ ===")
robots = [hero_robot, infantry_1, infantry_2, engineer]
for robot in robots:
    print(f"  {robot}")

print(f"\næ€»å…±åˆ›å»ºäº† {Robot.robot_count} ä¸ªæœºå™¨äºº")

# æ¼”ç¤ºåŸºæœ¬æ“ä½œ
print(f"\n=== åŸºæœ¬æ“ä½œæ¼”ç¤º ===")

# ç§»åŠ¨æµ‹è¯•
hero_robot.move(5, 3)
engineer.move(2, 1)

# æ”»å‡»æµ‹è¯•
print(f"\n--- æ”»å‡»æµ‹è¯• ---")
hero_robot.attack(infantry_1)     # è‹±é›„æ”»å‡»æ­¥å…µ
infantry_2.attack(hero_robot)     # æ­¥å…µæ”»å‡»è‹±é›„
engineer.attack(infantry_1)       # å·¥ç¨‹è¯•å›¾æ”»å‡»ï¼ˆåº”è¯¥å¤±è´¥ï¼‰

# å…‘çŸ¿æµ‹è¯•
print(f"\n--- ç‰¹æ®Šèƒ½åŠ›æµ‹è¯• ---")
engineer.mine_ore()               # å·¥ç¨‹å…‘çŸ¿
hero_robot.mine_ore()             # è‹±é›„è¯•å›¾å…‘çŸ¿ï¼ˆåº”è¯¥å¤±è´¥ï¼‰

# ä¿®å¤æµ‹è¯•
print(f"\n--- ä¿®å¤æµ‹è¯• ---")
hero_robot.repair(50)             # ä¿®å¤è‹±é›„
infantry_1.repair(100)            # ä¿®å¤æ­¥å…µ

# æŸ¥çœ‹æœ€ç»ˆçŠ¶æ€
print(f"\n=== æœ€ç»ˆçŠ¶æ€ ===")
for robot in robots:
    info = robot.get_info()
    print(f"  {info['name']}: {info['type']}, {info['hp']}HP, {info['damage']}æ”»å‡», çŠ¶æ€:{info['status']}")

print("\nâœ… åŸºç¡€ç±»å®šä¹‰å’Œå¯¹è±¡æ“ä½œå®Œæˆï¼")
```

### 2.1 å°è£… (Encapsulation)

å°è£…æ˜¯å°†æ•°æ®å’Œæ“ä½œæ•°æ®çš„æ–¹æ³•ç»‘å®šåœ¨ä¸€èµ·ï¼Œå¹¶éšè—å†…éƒ¨å®ç°ç»†èŠ‚çš„ç‰¹æ€§ã€‚

```python
# å°è£…å’Œç»§æ‰¿ç¤ºä¾‹ - åŸºäºçœŸå®RoboMasterè§„åˆ™

print("=== é¢å‘å¯¹è±¡ä¸‰å¤§ç‰¹æ€§æ¼”ç¤º ===")

# åŸºç±»ï¼šæœºå™¨äºº
class Robot:
    """åŸºç¡€æœºå™¨äººç±» - ç»Ÿä¸€çš„æœºå™¨äººæ¥å£"""
    
    def __init__(self, name, robot_type, hp):
        self.name = name
        self.robot_type = robot_type
        self._hp = hp  # å—ä¿æŠ¤å±æ€§ï¼ˆä½¿ç”¨ä¸‹åˆ’çº¿è¡¨ç¤ºï¼‰
        self._max_hp = hp
        self.position = (0, 0)
        self.status = "å¾…å‘½"
    
    @property
    def hp(self):
        """è¡€é‡å±æ€§ï¼ˆåªè¯»ï¼‰"""
        return self._hp
    
    @hp.setter 
    def hp(self, value):
        """è®¾ç½®è¡€é‡ï¼Œè‡ªåŠ¨é™åˆ¶åœ¨åˆç†èŒƒå›´å†…"""
        self._hp = max(0, min(value, self._max_hp))
        if self._hp == 0:
            self.status = "æŸå"
    
    @property
    def max_hp(self):
        """æœ€å¤§è¡€é‡å±æ€§ï¼ˆåªè¯»ï¼‰"""
        return self._max_hp
    
    def move(self, x, y):
        """ç§»åŠ¨æ–¹æ³•"""
        if self.status == "å¾…å‘½":
            self.position = (x, y)
            print(f"{self.name} ç§»åŠ¨åˆ° ({x}, {y})")
        else:
            print(f"âŒ {self.name} æ— æ³•ç§»åŠ¨ï¼ŒçŠ¶æ€: {self.status}")
    
    def attack(self, target):
        """åŸºç¡€æ”»å‡»æ–¹æ³•ï¼ˆå­ç±»åº”é‡å†™ï¼‰"""
        print(f"{self.name} å¯¹ {target} æ‰§è¡ŒåŸºç¡€æ”»å‡»")
        return 0
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        old_hp = self.hp
        self.hp -= damage
        print(f"ğŸ’¥ {self.name} å—åˆ° {damage} ä¼¤å®³: {old_hp} -> {self.hp}")

# æ´¾ç”Ÿç±»ï¼šè‹±é›„æœºå™¨äºº
class HeroRobot(Robot):
    """è‹±é›„æœºå™¨äººç±» - è¡€é‡åšï¼Œå¤§å¼¹ä¸¸æ”»å‡»"""
    
    def __init__(self, name):
        super().__init__(name, "Hero", 600)  # è‹±é›„è¡€é‡600
        self.weapon = "42mmå¤§å¼¹ä¸¸"
        self.armor_level = "é‡è£…ç”²"
        self.damage_per_shot = 10
    
    def attack(self, target):
        """é‡å†™æ”»å‡»æ–¹æ³• - å¤§å¼¹ä¸¸æ”»å‡»"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•æ”»å‡»ï¼ŒçŠ¶æ€: {self.status}")
            return 0
        
        print(f"ğŸ”¥ {self.name} ä½¿ç”¨ {self.weapon} æ”»å‡» {target.name}!")
        target.take_damage(self.damage_per_shot)
        return self.damage_per_shot
    
    def charge_attack(self, target):
        """è‹±é›„ç‰¹æœ‰æŠ€èƒ½ - è“„åŠ›æ”»å‡»"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•ä½¿ç”¨ç‰¹æ®Šæ”»å‡»")
            return 0
        
        charge_damage = self.damage_per_shot * 2  # è“„åŠ›ä¼¤å®³ç¿»å€
        print(f"âš¡ {self.name} å‘åŠ¨è“„åŠ›æ”»å‡»!")
        target.take_damage(charge_damage)
        return charge_damage

# æ´¾ç”Ÿç±»ï¼šæ­¥å…µæœºå™¨äºº  
class InfantryRobot(Robot):
    """æ­¥å…µæœºå™¨äººç±» - æ ‡å‡†è¡€é‡ï¼Œå°å¼¹ä¸¸æ”»å‡»"""
    
    def __init__(self, name):
        super().__init__(name, "Infantry", 200)  # æ­¥å…µè¡€é‡200
        self.weapon = "17mmå°å¼¹ä¸¸"
        self.mobility = "é«˜æœºåŠ¨æ€§"
        self.damage_per_shot = 1
    
    def attack(self, target):
        """é‡å†™æ”»å‡»æ–¹æ³• - å°å¼¹ä¸¸æ”»å‡»"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•æ”»å‡»ï¼ŒçŠ¶æ€: {self.status}")
            return 0
        
        print(f"ğŸ”« {self.name} ä½¿ç”¨ {self.weapon} æ”»å‡» {target.name}!")
        target.take_damage(self.damage_per_shot)
        return self.damage_per_shot
    
    def rapid_fire(self, target, shots=3):
        """æ­¥å…µç‰¹æœ‰æŠ€èƒ½ - è¿å‘å°„å‡»"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•ä½¿ç”¨è¿å‘å°„å‡»")
            return 0
        
        total_damage = 0
        print(f"ğŸ”«ğŸ”«ğŸ”« {self.name} å¯¹ {target.name} è¿›è¡Œ {shots} è¿å‘å°„å‡»!")
        for i in range(shots):
            target.take_damage(self.damage_per_shot)
            total_damage += self.damage_per_shot
        return total_damage

# æ´¾ç”Ÿç±»ï¼šå·¥ç¨‹æœºå™¨äºº
class EngineerRobot(Robot):
    """å·¥ç¨‹æœºå™¨äººç±» - ä¸­ç­‰è¡€é‡ï¼Œæ— æ”»å‡»èƒ½åŠ›ï¼Œå¯å…‘çŸ¿"""
    
    def __init__(self, name):
        super().__init__(name, "Engineer", 250)  # å·¥ç¨‹è¡€é‡250
        self.tool = "å…‘çŸ¿è£…ç½®"
        self.ore_capacity = 0
        self.max_ore_capacity = 100
    
    def attack(self, target):
        """å·¥ç¨‹æœºå™¨äººæ— æ”»å‡»èƒ½åŠ›"""
        print(f"âŒ {self.name} æ˜¯å·¥ç¨‹æœºå™¨äººï¼Œæ— æ³•æ”»å‡»")
        return 0
    
    def mine_ore(self, amount=20):
        """å…‘çŸ¿æ“ä½œ"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•å…‘çŸ¿ï¼ŒçŠ¶æ€: {self.status}")
            return 0
        
        if self.ore_capacity >= self.max_ore_capacity:
            print(f"âŒ {self.name} çŸ¿çŸ³å®¹é‡å·²æ»¡")
            return 0
        
        actual_amount = min(amount, self.max_ore_capacity - self.ore_capacity)
        self.ore_capacity += actual_amount
        print(f"â›ï¸ {self.name} å…‘å– {actual_amount} å•ä½çŸ¿çŸ³ (æ€»é‡: {self.ore_capacity}/{self.max_ore_capacity})")
        return actual_amount
    
    def repair_ally(self, target, heal_amount=50):
        """ä¿®å¤ç›Ÿå‹"""
        if self.status != "å¾…å‘½":
            print(f"âŒ {self.name} æ— æ³•ä¿®å¤ï¼ŒçŠ¶æ€: {self.status}")
            return 0
        
        if target.status == "æŸå":
            print(f"âŒ {target.name} å·²æŸåï¼Œæ— æ³•ä¿®å¤")
            return 0
        
        old_hp = target.hp
        target.hp = min(target.max_hp, target.hp + heal_amount)
        actual_heal = target.hp - old_hp
        print(f"ğŸ”§ {self.name} ä¿®å¤ {target.name}: {old_hp} -> {target.hp} (+{actual_heal})")
        return actual_heal

# åˆ›å»ºæœºå™¨äººé˜Ÿä¼
print("\n=== åˆ›å»ºä¸“ä¸šæœºå™¨äººé˜Ÿä¼ ===")
hero = HeroRobot("Hero-Alpha")
infantry1 = InfantryRobot("Infantry-Beta")  
infantry2 = InfantryRobot("Infantry-Gamma")
engineer = EngineerRobot("Engineer-Delta")

robots = [hero, infantry1, infantry2, engineer]

# æ˜¾ç¤ºåˆå§‹çŠ¶æ€
print(f"\n=== é˜Ÿä¼åˆå§‹çŠ¶æ€ ===")
for robot in robots:
    print(f"  {robot.name}: {robot.robot_type}, {robot.hp}/{robot.max_hp}HP")

# æ¼”ç¤ºä¸åŒæ”»å‡»æ–¹å¼
print(f"\n=== æˆ˜æ–—æ¼”ç¤º ===")

# è‹±é›„æ”»å‡»æ­¥å…µ
hero.attack(infantry1)
hero.charge_attack(infantry1)  # ç‰¹æ®Šæ”»å‡»

# æ­¥å…µåå‡»
infantry1.attack(hero)
infantry1.rapid_fire(hero, 5)  # è¿å‘æ”»å‡»

# å·¥ç¨‹æœºå™¨äººå°è¯•æ”»å‡»ï¼ˆåº”è¯¥å¤±è´¥ï¼‰
engineer.attack(hero)

# æ¼”ç¤ºç‰¹æ®Šèƒ½åŠ›
print(f"\n=== ç‰¹æ®Šèƒ½åŠ›æ¼”ç¤º ===")
engineer.mine_ore(30)
engineer.mine_ore(50)
engineer.repair_ally(infantry1, 80)

# æ˜¾ç¤ºæœ€ç»ˆçŠ¶æ€
print(f"\n=== æˆ˜æ–—åçŠ¶æ€ ===")
for robot in robots:
    status_info = f"{robot.name}: {robot.robot_type}, {robot.hp}/{robot.max_hp}HP"
    if hasattr(robot, 'ore_capacity'):
        status_info += f", çŸ¿çŸ³: {robot.ore_capacity}"
    print(f"  {status_info}")

print(f"\nâœ… ç»§æ‰¿å’Œå¤šæ€æ¼”ç¤ºå®Œæˆï¼")
```

### 3.1 å¼‚å¸¸çš„åŸºæœ¬æ¦‚å¿µ

å¼‚å¸¸æ˜¯ç¨‹åºè¿è¡Œæ—¶å‘ç”Ÿçš„é”™è¯¯ï¼ŒPythonæä¾›äº†å®Œå–„çš„å¼‚å¸¸å¤„ç†æœºåˆ¶æ¥å¤„ç†è¿™äº›é”™è¯¯ã€‚

```python
# å¼‚å¸¸å¤„ç†è¯¦è§£

print("=== å¼‚å¸¸å¤„ç†æœºåˆ¶ ===")

# è‡ªå®šä¹‰å¼‚å¸¸ç±»
class RobotError(Exception):
    """æœºå™¨äººç›¸å…³å¼‚å¸¸çš„åŸºç±»"""
    pass

class RobotDestroyedException(RobotError):
    """æœºå™¨äººå·²æŸåå¼‚å¸¸"""
    pass

class InsufficientEnergyException(RobotError):
    """èƒ½é‡ä¸è¶³å¼‚å¸¸"""
    pass

class InvalidPositionException(RobotError):
    """æ— æ•ˆä½ç½®å¼‚å¸¸"""
    pass

class RobotControlSystem:
    """æœºå™¨äººæ§åˆ¶ç³»ç»Ÿ - å¼‚å¸¸å¤„ç†ç¤ºä¾‹"""
    
    def __init__(self):
        self.robots = {}
        self.field_size = (1000, 1000)
    
    def add_robot(self, robot_id, name, robot_type):
        """æ·»åŠ æœºå™¨äºº"""
        try:
            if robot_id in self.robots:
                raise ValueError(f"æœºå™¨äººID {robot_id} å·²å­˜åœ¨")
            
            hp_map = {"Hero": 400, "Infantry": 200, "Engineer": 250}
            if robot_type not in hp_map:
                raise ValueError(f"ä¸æ”¯æŒçš„æœºå™¨äººç±»å‹: {robot_type}")
            
            self.robots[robot_id] = {
                "name": name,
                "type": robot_type,
                "hp": hp_map[robot_type],
                "max_hp": hp_map[robot_type],
                "energy": 100,
                "position": (0, 0),
                "status": "æ­£å¸¸"
            }
            
            print(f"âœ… æˆåŠŸæ·»åŠ æœºå™¨äºº: {name} (ID: {robot_id})")
            
        except ValueError as e:
            print(f"âŒ æ·»åŠ æœºå™¨äººå¤±è´¥: {e}")
            raise
    
    def move_robot(self, robot_id, x, y):
        """ç§»åŠ¨æœºå™¨äºº"""
        try:
            # æ£€æŸ¥æœºå™¨äººæ˜¯å¦å­˜åœ¨
            if robot_id not in self.robots:
                raise KeyError(f"æœºå™¨äººID {robot_id} ä¸å­˜åœ¨")
            
            robot = self.robots[robot_id]
            
            # æ£€æŸ¥æœºå™¨äººçŠ¶æ€
            if robot["status"] == "æŸå":
                raise RobotDestroyedException(f"æœºå™¨äºº {robot['name']} å·²æŸåï¼Œæ— æ³•ç§»åŠ¨")
            
            # æ£€æŸ¥ä½ç½®æœ‰æ•ˆæ€§
            if not (0 <= x <= self.field_size[0] and 0 <= y <= self.field_size[1]):
                raise InvalidPositionException(f"ä½ç½® ({x}, {y}) è¶…å‡ºåœºåœ°èŒƒå›´")
            
            # æ£€æŸ¥èƒ½é‡
            energy_cost = 10
            if robot["energy"] < energy_cost:
                raise InsufficientEnergyException(f"èƒ½é‡ä¸è¶³ï¼Œéœ€è¦ {energy_cost}ï¼Œå½“å‰ {robot['energy']}")
            
            # æ‰§è¡Œç§»åŠ¨
            old_pos = robot["position"]
            robot["position"] = (x, y)
            robot["energy"] -= energy_cost
            
            print(f"âœ… {robot['name']} ä» {old_pos} ç§»åŠ¨åˆ° ({x}, {y})ï¼Œæ¶ˆè€—èƒ½é‡ {energy_cost}")
            
        except (KeyError, RobotDestroyedException, InvalidPositionException, InsufficientEnergyException) as e:
            print(f"âŒ ç§»åŠ¨å¤±è´¥: {e}")
            return False
        except Exception as e:
            print(f"âŒ æ„å¤–é”™è¯¯: {e}")
            return False
        
        return True
    
    def attack_robot(self, attacker_id, target_id):
        """æœºå™¨äººæ”»å‡»"""
        try:
            # æ£€æŸ¥æ”»å‡»è€…å’Œç›®æ ‡
            if attacker_id not in self.robots:
                raise KeyError(f"æ”»å‡»è€… {attacker_id} ä¸å­˜åœ¨")
            if target_id not in self.robots:
                raise KeyError(f"ç›®æ ‡ {target_id} ä¸å­˜åœ¨")
            
            attacker = self.robots[attacker_id]
            target = self.robots[target_id]
            
            # æ£€æŸ¥æ”»å‡»è€…çŠ¶æ€
            if attacker["status"] == "æŸå":
                raise RobotDestroyedException(f"æ”»å‡»è€… {attacker['name']} å·²æŸå")
            
            # æ£€æŸ¥ç›®æ ‡çŠ¶æ€
            if target["status"] == "æŸå":
                raise RobotDestroyedException(f"ç›®æ ‡ {target['name']} å·²æŸå")
            
            # æ£€æŸ¥èƒ½é‡
            if attacker["energy"] < 20:
                raise InsufficientEnergyException("æ”»å‡»éœ€è¦20ç‚¹èƒ½é‡")
            
            # æ‰§è¡Œæ”»å‡»
            damage = 50
            attacker["energy"] -= 20
            target["hp"] -= damage
            
            print(f"ğŸ’¥ {attacker['name']} æ”»å‡» {target['name']}ï¼Œé€ æˆ {damage} ä¼¤å®³")
            
            # æ£€æŸ¥ç›®æ ‡æ˜¯å¦è¢«æ‘§æ¯
            if target["hp"] <= 0:
                target["hp"] = 0
                target["status"] = "æŸå"
                print(f"ğŸ’€ {target['name']} è¢«æ‘§æ¯!")
            
        except Exception as e:
            print(f"âŒ æ”»å‡»å¤±è´¥: {e}")
            return False
        
        return True
    
    def get_robot_status(self, robot_id):
        """è·å–æœºå™¨äººçŠ¶æ€"""
        try:
            if robot_id not in self.robots:
                raise KeyError(f"æœºå™¨äººID {robot_id} ä¸å­˜åœ¨")
            
            return self.robots[robot_id].copy()
            
        except KeyError as e:
            print(f"âŒ æŸ¥è¯¢å¤±è´¥: {e}")
            return None

# å¼‚å¸¸å¤„ç†æµ‹è¯•
print("\\n=== å¼‚å¸¸å¤„ç†æµ‹è¯• ===")

system = RobotControlSystem()

# æµ‹è¯•æ­£å¸¸æ“ä½œ
try:
    system.add_robot("R001", "Hero-1", "Hero")
    system.add_robot("R002", "Infantry-1", "Infantry")
    print("\\næ­£å¸¸æ“ä½œå®Œæˆ")
except Exception as e:
    print(f"æ“ä½œå¤±è´¥: {e}")

# æµ‹è¯•å¼‚å¸¸æƒ…å†µ
print(f"\\n=== å¼‚å¸¸æƒ…å†µæµ‹è¯• ===")

# 1. é‡å¤IDå¼‚å¸¸
try:
    system.add_robot("R001", "Hero-2", "Hero")
except ValueError:
    print("âœ… æˆåŠŸæ•è·é‡å¤IDå¼‚å¸¸")

# 2. æ— æ•ˆæœºå™¨äººç±»å‹
try:
    system.add_robot("R003", "Unknown", "Tank")
except ValueError:
    print("âœ… æˆåŠŸæ•è·æ— æ•ˆç±»å‹å¼‚å¸¸")

# 3. ç§»åŠ¨åˆ°æ— æ•ˆä½ç½®
try:
    system.move_robot("R001", 2000, 3000)
except:
    print("âœ… æˆåŠŸæ•è·æ— æ•ˆä½ç½®å¼‚å¸¸")

# 4. èƒ½é‡ä¸è¶³ï¼ˆé€šè¿‡å¤šæ¬¡ç§»åŠ¨æ¶ˆè€—èƒ½é‡ï¼‰
print(f"\\n=== èƒ½é‡æ¶ˆè€—æµ‹è¯• ===")
for i in range(12):  # å°è¯•ç§»åŠ¨12æ¬¡ï¼Œæ¯æ¬¡æ¶ˆè€—10èƒ½é‡
    success = system.move_robot("R001", i*10, i*10)
    if not success:
        print(f"ç¬¬{i+1}æ¬¡ç§»åŠ¨å¤±è´¥")
        break

# 5. æ”»å‡»å·²æŸåçš„æœºå™¨äºº
print(f"\\n=== æˆ˜æ–—æµ‹è¯• ===")
# å…ˆæ”»å‡»ç›´åˆ°ç›®æ ‡è¢«æ‘§æ¯
for i in range(10):  # å¤šæ¬¡æ”»å‡»
    success = system.attack_robot("R001", "R002")
    if not success:
        break

# å°è¯•æ”»å‡»å·²æŸåçš„æœºå™¨äºº
system.attack_robot("R001", "R002")

print("\\nâœ… å¼‚å¸¸å¤„ç†å­¦ä¹ å®Œæˆï¼")
```

## ç¬¬å››éƒ¨åˆ†ï¼šç»¼åˆå®è·µé¡¹ç›® - æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨

è®©æˆ‘ä»¬åˆ›å»ºä¸€ä¸ªå®Œæ•´çš„æœºå™¨äººæ§åˆ¶ç³»ç»Ÿï¼Œç»¼åˆè¿ç”¨æ‰€æœ‰å­¦åˆ°çš„çŸ¥è¯†ã€‚

```python
# ç»¼åˆå®è·µé¡¹ç›®ï¼šæœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨ - åŸºäºçœŸå®RoboMasterè§„åˆ™

import random
from datetime import datetime
from enum import Enum

print("ğŸ¤– RoboMaster æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨")
print("=" * 60)

class RobotState(Enum):
    """æœºå™¨äººçŠ¶æ€æšä¸¾"""
    STANDBY = "å¾…å‘½"
    ACTIVE = "æ¿€æ´»"
    MOVING = "ç§»åŠ¨ä¸­" 
    FIGHTING = "æˆ˜æ–—ä¸­"
    MINING = "å…‘çŸ¿ä¸­"
    DESTROYED = "å·²æŸå"

class BattleRobot:
    """æˆ˜æ–—æœºå™¨äººåŸºç±» - å®ç°çœŸå®RoboMasterè§„åˆ™"""
    
    def __init__(self, robot_id, name, robot_type):
        self.robot_id = robot_id
        self.name = name
        self.robot_type = robot_type
        self.state = RobotState.STANDBY
        self.position = (0, 0)
        self.energy = 100
        self.max_energy = 100
        
        # æ ¹æ®æœºå™¨äººç±»å‹è®¾ç½®å±æ€§ï¼ˆçœŸå®è§„åˆ™ï¼‰
        if robot_type == "Hero":
            self.hp = 600
            self.max_hp = 600
            self.damage = 10  # å¤§å¼¹ä¸¸
            self.energy_per_move = 8
            self.energy_per_attack = 15
            self.can_attack = True
            self.can_mine = False
        elif robot_type == "Infantry":
            self.hp = 200
            self.max_hp = 200
            self.damage = 1   # å°å¼¹ä¸¸
            self.energy_per_move = 5
            self.energy_per_attack = 8
            self.can_attack = True
            self.can_mine = False
        elif robot_type == "Engineer":
            self.hp = 250
            self.max_hp = 250
            self.damage = 0   # æ— æ”»å‡»èƒ½åŠ›
            self.energy_per_move = 6
            self.energy_per_attack = 0
            self.can_attack = False
            self.can_mine = True
            self.ore_count = 0
            self.max_ore = 100
        else:
            raise ValueError(f"æœªçŸ¥çš„æœºå™¨äººç±»å‹: {robot_type}")
        
        self.battle_log = []
        
    def log_action(self, action):
        """è®°å½•æ“ä½œæ—¥å¿—"""
        timestamp = datetime.now().strftime("%H:%M:%S")
        log_entry = f"[{timestamp}] {self.name}: {action}"
        self.battle_log.append(log_entry)
        print(log_entry)
    
    def move(self, x, y):
        """ç§»åŠ¨æœºå™¨äºº"""
        if self.state == RobotState.DESTROYED:
            self.log_action("âŒ å·²æŸåï¼Œæ— æ³•ç§»åŠ¨")
            return False
        
        if self.energy < self.energy_per_move:
            self.log_action(f"âŒ èƒ½é‡ä¸è¶³ï¼Œéœ€è¦ {self.energy_per_move} ç‚¹èƒ½é‡")
            return False
        
        old_pos = self.position
        self.position = (x, y)
        self.energy -= self.energy_per_move
        self.state = RobotState.MOVING
        
        self.log_action(f"ç§»åŠ¨: {old_pos} -> {self.position}, æ¶ˆè€— {self.energy_per_move} èƒ½é‡")
        return True
    
    def attack(self, target):
        """æ”»å‡»ç›®æ ‡"""
        if not self.can_attack:
            self.log_action(f"âŒ {self.robot_type} æ— æ”»å‡»èƒ½åŠ›")
            return False
        
        if self.state == RobotState.DESTROYED:
            self.log_action("âŒ å·²æŸåï¼Œæ— æ³•æ”»å‡»")
            return False
        
        if self.energy < self.energy_per_attack:
            self.log_action(f"âŒ èƒ½é‡ä¸è¶³ï¼Œéœ€è¦ {self.energy_per_attack} ç‚¹èƒ½é‡")
            return False
        
        if target.state == RobotState.DESTROYED:
            self.log_action(f"âŒ ç›®æ ‡ {target.name} å·²è¢«æ‘§æ¯")
            return False
        
        # æ‰§è¡Œæ”»å‡»
        self.energy -= self.energy_per_attack
        self.state = RobotState.FIGHTING
        
        # è®¡ç®—ä¼¤å®³ï¼ˆå¯èƒ½æœ‰æš´å‡»ï¼‰
        damage = self.damage
        is_critical = random.random() < 0.1  # 10%æš´å‡»ç‡
        if is_critical:
            damage *= 2
            self.log_action(f"ğŸ”¥ æš´å‡»! å¯¹ {target.name} é€ æˆ {damage} ç‚¹ä¼¤å®³")
        else:
            self.log_action(f"ğŸ”« å¯¹ {target.name} é€ æˆ {damage} ç‚¹ä¼¤å®³")
        
        # ç›®æ ‡å—ä¼¤
        target.take_damage(damage)
        return True
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        old_hp = self.hp
        self.hp = max(0, self.hp - damage)
        
        if self.hp == 0:
            self.state = RobotState.DESTROYED
            self.log_action(f"ğŸ’€ å—åˆ° {damage} ä¼¤å®³ ({old_hp} -> {self.hp})ï¼Œæœºå™¨äººè¢«æ‘§æ¯!")
        else:
            self.log_action(f"ğŸ’¥ å—åˆ° {damage} ä¼¤å®³ ({old_hp} -> {self.hp})")
    
    def mine_ore(self):
        """å…‘çŸ¿æ“ä½œï¼ˆä»…å·¥ç¨‹æœºå™¨äººï¼‰"""
        if not self.can_mine:
            self.log_action(f"âŒ {self.robot_type} æ— æ³•å…‘çŸ¿")
            return False
        
        if self.state == RobotState.DESTROYED:
            self.log_action("âŒ å·²æŸåï¼Œæ— æ³•å…‘çŸ¿")
            return False
        
        if self.ore_count >= self.max_ore:
            self.log_action("âŒ çŸ¿çŸ³å®¹é‡å·²æ»¡")
            return False
        
        # å…‘çŸ¿æ“ä½œ
        ore_gained = random.randint(10, 30)
        self.ore_count = min(self.max_ore, self.ore_count + ore_gained)
        self.state = RobotState.MINING
        
        self.log_action(f"â›ï¸ å…‘å– {ore_gained} çŸ¿çŸ³ï¼Œæ€»è®¡: {self.ore_count}/{self.max_ore}")
        return True
    
    def repair(self, heal_amount):
        """ä¿®å¤ï¼ˆéœ€è¦æ¶ˆè€—çŸ¿çŸ³ï¼‰"""
        if self.state == RobotState.DESTROYED:
            self.log_action("âŒ æœºå™¨äººå·²å®Œå…¨æŸåï¼Œæ— æ³•ä¿®å¤")
            return False
        
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + heal_amount)
        actual_heal = self.hp - old_hp
        
        self.log_action(f"ğŸ”§ ä¿®å¤ {actual_heal} ç‚¹è¡€é‡ ({old_hp} -> {self.hp})")
        return True
    
    def recharge(self, energy_amount=50):
        """è¡¥å……èƒ½é‡"""
        old_energy = self.energy
        self.energy = min(self.max_energy, self.energy + energy_amount)
        actual_recharge = self.energy - old_energy
        
        self.log_action(f"ğŸ”‹ è¡¥å…… {actual_recharge} ç‚¹èƒ½é‡ ({old_energy} -> {self.energy})")
        return actual_recharge
    
    def get_status(self):
        """è·å–çŠ¶æ€ä¿¡æ¯"""
        status = {
            "name": self.name,
            "type": self.robot_type,
            "hp": f"{self.hp}/{self.max_hp}",
            "energy": f"{self.energy}/{self.max_energy}",
            "damage": self.damage,
            "position": self.position,
            "state": self.state.value
        }
        
        if hasattr(self, 'ore_count'):
            status["ore"] = f"{self.ore_count}/{self.max_ore}"
        
        return status

class BattleSimulator:
    """æˆ˜æ–—æ¨¡æ‹Ÿå™¨"""
    
    def __init__(self):
        self.robots = {}
        self.battle_round = 0
        self.battle_log = []
    
    def add_robot(self, robot_id, name, robot_type):
        """æ·»åŠ æœºå™¨äºº"""
        if robot_id in self.robots:
            print(f"âŒ æœºå™¨äººID {robot_id} å·²å­˜åœ¨")
            return False
        
        robot = BattleRobot(robot_id, name, robot_type)
        self.robots[robot_id] = robot
        print(f"âœ… æ·»åŠ  {robot_type} æœºå™¨äºº: {name} (ID: {robot_id})")
        return True
    
    def get_robot(self, robot_id):
        """è·å–æœºå™¨äºº"""
        return self.robots.get(robot_id)
    
    def display_all_robots(self):
        """æ˜¾ç¤ºæ‰€æœ‰æœºå™¨äººçŠ¶æ€"""
        print(f"\nğŸ“Š æœºå™¨äººçŠ¶æ€æ€»è§ˆ (ç¬¬ {self.battle_round} è½®)")
        print("-" * 80)
        
        for robot_id, robot in self.robots.items():
            status = robot.get_status()
            ore_info = f", çŸ¿çŸ³: {status['ore']}" if 'ore' in status else ""
            print(f"  {robot_id}: {status['name']} ({status['type']}) - "
                  f"HP: {status['hp']}, èƒ½é‡: {status['energy']}, "
                  f"ä¼¤å®³: {status['damage']}, çŠ¶æ€: {status['state']}{ore_info}")
    
    def simulate_battle(self):
        """æ¨¡æ‹Ÿä¸€è½®æˆ˜æ–—"""
        self.battle_round += 1
        print(f"\nâš”ï¸ === ç¬¬ {self.battle_round} è½®æˆ˜æ–— ===")
        
        active_robots = [r for r in self.robots.values() if r.state != RobotState.DESTROYED]
        
        if len(active_robots) < 2:
            print("æˆ˜æ–—ç»“æŸï¼Œå‰©ä½™æœºå™¨äººä¸è¶³2ä¸ª")
            return False
        
        # éšæœºé€‰æ‹©æ”»å‡»è€…å’Œç›®æ ‡
        attacker = random.choice([r for r in active_robots if r.can_attack and r.energy >= r.energy_per_attack])
        if not attacker:
            print("æ²¡æœ‰å¯ç”¨çš„æ”»å‡»æœºå™¨äºº")
            return False
        
        targets = [r for r in active_robots if r != attacker and r.state != RobotState.DESTROYED]
        if not targets:
            print("æ²¡æœ‰å¯æ”»å‡»çš„ç›®æ ‡")
            return False
        
        target = random.choice(targets)
        
        # æ‰§è¡Œæ”»å‡»
        attacker.attack(target)
        
        # å·¥ç¨‹æœºå™¨äººå¯èƒ½è¿›è¡Œå…‘çŸ¿
        engineers = [r for r in active_robots if r.robot_type == "Engineer" and r.state != RobotState.DESTROYED]
        for engineer in engineers:
            if random.random() < 0.3:  # 30%æ¦‚ç‡å…‘çŸ¿
                engineer.mine_ore()
        
        return True
    
    def run_simulation(self, rounds=5):
        """è¿è¡Œå®Œæ•´æ¨¡æ‹Ÿ"""
        print(f"\nğŸ® å¼€å§‹ {rounds} è½®æˆ˜æ–—æ¨¡æ‹Ÿ")
        
        for round_num in range(rounds):
            if not self.simulate_battle():
                break
            
            # æ¯è½®åè¡¥å……å°‘é‡èƒ½é‡
            for robot in self.robots.values():
                if robot.state != RobotState.DESTROYED:
                    robot.recharge(20)
            
            self.display_all_robots()
            
        print(f"\nğŸ æˆ˜æ–—æ¨¡æ‹Ÿç»“æŸ")
        
        # ç»Ÿè®¡æœ€ç»ˆç»“æœ
        survivors = [r for r in self.robots.values() if r.state != RobotState.DESTROYED]
        destroyed = [r for r in self.robots.values() if r.state == RobotState.DESTROYED]
        
        print(f"\nğŸ“ˆ æˆ˜æ–—ç»Ÿè®¡:")
        print(f"  å­˜æ´»æœºå™¨äºº: {len(survivors)}")
        print(f"  è¢«æ‘§æ¯æœºå™¨äºº: {len(destroyed)}")
        
        if survivors:
            print(f"  å­˜æ´»è€…:")
            for robot in survivors:
                print(f"    {robot.name} ({robot.robot_type}): {robot.hp}/{robot.max_hp} HP")

# åˆ›å»ºæˆ˜æ–—æ¨¡æ‹Ÿå™¨å¹¶è¿è¡Œæ¼”ç¤º
print("\n=== åˆ›å»ºæœºå™¨äººé˜Ÿä¼ ===")
simulator = BattleSimulator()

# æ·»åŠ ä¸åŒç±»å‹çš„æœºå™¨äºº
simulator.add_robot("H001", "Hero-Alpha", "Hero")
simulator.add_robot("I001", "Infantry-Beta", "Infantry")
simulator.add_robot("I002", "Infantry-Gamma", "Infantry")
simulator.add_robot("E001", "Engineer-Delta", "Engineer")

# æ˜¾ç¤ºåˆå§‹çŠ¶æ€
simulator.display_all_robots()

# æ‰‹åŠ¨æ¼”ç¤ºå‡ ä¸ªæ“ä½œ
print(f"\n=== æ‰‹åŠ¨æ“ä½œæ¼”ç¤º ===")
hero = simulator.get_robot("H001")
infantry1 = simulator.get_robot("I001")
engineer = simulator.get_robot("E001")

# ç§»åŠ¨æµ‹è¯•
hero.move(10, 5)
infantry1.move(15, 8)

# æ”»å‡»æµ‹è¯•
hero.attack(infantry1)
infantry1.attack(hero)

# å·¥ç¨‹æœºå™¨äººå…‘çŸ¿
engineer.mine_ore()
engineer.mine_ore()

# ä¿®å¤æµ‹è¯•
engineer_heal_amount = 30
infantry1.repair(engineer_heal_amount)

# è¿è¡Œè‡ªåŠ¨æˆ˜æ–—æ¨¡æ‹Ÿ
simulator.run_simulation(8)

print(f"\n=== é¡¹ç›®çŸ¥è¯†ç‚¹æ€»ç»“ ===")
print("âœ… ç±»çš„è®¾è®¡ä¸ç»§æ‰¿")
print("âœ… æšä¸¾ç±»å‹çš„ä½¿ç”¨")
print("âœ… å¼‚å¸¸å¤„ç†æœºåˆ¶")
print("âœ… éšæœºæ•°å’Œæ¦‚ç‡è®¡ç®—")
print("âœ… æ—¥æœŸæ—¶é—´å¤„ç†")
print("âœ… å¤æ‚æ•°æ®ç»“æ„ç®¡ç†")
print("âœ… é¢å‘å¯¹è±¡è®¾è®¡åŸåˆ™")
print("âœ… çœŸå®ä¸šåŠ¡é€»è¾‘å»ºæ¨¡")

print(f"\nğŸ‰ æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨é¡¹ç›®å®Œæˆ!")
print("è¿™ä¸ªé¡¹ç›®å±•ç¤ºäº†å¦‚ä½•ç”¨é¢å‘å¯¹è±¡ç¼–ç¨‹æ„å»ºå¤æ‚çš„çœŸå®ç³»ç»Ÿï¼")
```
