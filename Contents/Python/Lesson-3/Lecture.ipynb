é¢å‘å¯¹è±¡ç¼–ç¨‹æ˜¯ä¸€ç§ç¼–ç¨‹èŒƒå¼ï¼Œå®ƒå°†ç°å®ä¸–ç•Œçš„äº‹ç‰©æŠ½è±¡ä¸ºç¨‹åºä¸­çš„å¯¹è±¡ã€‚

- **å¯¹è±¡ (Object)**: ç°å®ä¸–ç•Œå®ä½“çš„æŠ½è±¡ï¼ŒåŒ…å«æ•°æ®å’Œæ–¹æ³•
- **ç±» (Class)**: å¯¹è±¡çš„æ¨¡æ¿æˆ–è“å›¾ï¼Œå®šä¹‰å¯¹è±¡çš„ç»“æ„
- **å®ä¾‹ (Instance)**: ç±»çš„å…·ä½“å®ç°ï¼Œä¸€ä¸ªçœŸå®å­˜åœ¨çš„å¯¹è±¡

**åœ¨ RoboMaster ä¸­çš„å¯¹è±¡æ€ç»´ï¼š**

- æœºå™¨äººç±»å‹ï¼ˆHeroã€Infantryç­‰ï¼‰ â†’ ç±»

```python
# ç¬¬ä¸€ä¸ªå®Œæ•´çš„ç±»å®šä¹‰

print("=== é¢å‘å¯¹è±¡ç¼–ç¨‹åŸºç¡€ ===")

class Robot:
    """æœºå™¨äººç±» - è¿™æ˜¯æˆ‘ä»¬çš„ç¬¬ä¸€ä¸ªå®Œæ•´ç±»å®šä¹‰"""
    
    # ç±»å±æ€§ï¼ˆæ‰€æœ‰å®ä¾‹å…±äº«ï¼‰
    game_version = "2024"
    robot_count = 0
    
    def __init__(self, name, robot_type, hp=200):
        """
        æ„é€ å‡½æ•°ï¼šåˆ›å»ºæœºå™¨äººå®ä¾‹æ—¶è‡ªåŠ¨è°ƒç”¨
        
        Args:
            name (str): æœºå™¨äººåç§°
            robot_type (str): æœºå™¨äººç±»å‹
            hp (int): åˆå§‹ç”Ÿå‘½å€¼
        """
        # å®ä¾‹å±æ€§ï¼ˆæ¯ä¸ªå®ä¾‹ç‹¬æœ‰ï¼‰
        self.name = name
        self.robot_type = robot_type
        self.hp = hp
        self.max_hp = hp
        self.position = (0, 0)
        self.status = "å¾…å‘½"
        
        # æ›´æ–°ç±»å±æ€§
        Robot.robot_count += 1
        
        print(f"âœ… åˆ›å»ºæœºå™¨äºº: {name} ({robot_type})")
    
    def __str__(self):
        """å­—ç¬¦ä¸²è¡¨ç¤º - ç”¨äºprint()"""
        return f"Robot({self.name}, {self.robot_type}, HP:{self.hp}/{self.max_hp})"
    
    def __repr__(self):
        """å¼€å‘è€…è¡¨ç¤º - ç”¨äºè°ƒè¯•"""
        return f"Robot('{self.name}', '{self.robot_type}', {self.hp})"
    
    # å®ä¾‹æ–¹æ³•
    def move(self, x, y):
        """ç§»åŠ¨æœºå™¨äººåˆ°æŒ‡å®šä½ç½®"""
        self.position = (x, y)
        print(f"{self.name} ç§»åŠ¨åˆ°ä½ç½® ({x}, {y})")
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        old_hp = self.hp
        self.hp = max(0, self.hp - damage)
        print(f"{self.name} å—åˆ° {damage} ç‚¹ä¼¤å®³: {old_hp} -> {self.hp}")
        
        if self.hp == 0:
            self.status = "æŸå"
            print(f"ğŸ’¥ {self.name} å·²è¢«æ‘§æ¯!")
    
    def repair(self, heal_amount):
        """ä¿®å¤æœºå™¨äºº"""
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + heal_amount)
        print(f"ğŸ”§ {self.name} ä¿®å¤ {heal_amount} ç‚¹: {old_hp} -> {self.hp}")
    
    def get_status(self):
        """è·å–æœºå™¨äººçŠ¶æ€ä¿¡æ¯"""
        hp_percent = (self.hp / self.max_hp) * 100
        return {
            "name": self.name,
            "type": self.robot_type,
            "hp": self.hp,
            "max_hp": self.max_hp,
            "hp_percent": hp_percent,
            "position": self.position,
            "status": self.status
        }
    
    # ç±»æ–¹æ³•
    @classmethod
    def get_robot_count(cls):
        """è·å–æœºå™¨äººæ€»æ•°"""
        return cls.robot_count
    
    @classmethod
    def create_hero(cls, name):
        """å·¥å‚æ–¹æ³•ï¼šåˆ›å»ºHeroç±»å‹æœºå™¨äºº"""
        return cls(name, "Hero", 400)
    
    @classmethod
    def create_infantry(cls, name):
        """å·¥å‚æ–¹æ³•ï¼šåˆ›å»ºInfantryç±»å‹æœºå™¨äºº"""
        return cls(name, "Infantry", 200)
    
    # é™æ€æ–¹æ³•
    @staticmethod
    def calculate_distance(pos1, pos2):
        """è®¡ç®—ä¸¤ç‚¹é—´è·ç¦»"""
        return ((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)**0.5
    
    @staticmethod
    def is_valid_position(x, y, field_width=1000, field_height=1000):
        """éªŒè¯ä½ç½®æ˜¯å¦åœ¨æœ‰æ•ˆèŒƒå›´å†…"""
        return 0 <= x <= field_width and 0 <= y <= field_height

# åˆ›å»ºå’Œä½¿ç”¨æœºå™¨äººå¯¹è±¡
print(f"\\n=== åˆ›å»ºæœºå™¨äººå®ä¾‹ ===")

# ä½¿ç”¨æ„é€ å‡½æ•°åˆ›å»º
hero = Robot("Hero-1", "Hero", 400)
infantry = Robot("Infantry-1", "Infantry", 200)
engineer = Robot("Engineer-1", "Engineer", 250)

# ä½¿ç”¨å·¥å‚æ–¹æ³•åˆ›å»º
hero2 = Robot.create_hero("Hero-2")
infantry2 = Robot.create_infantry("Infantry-2")

print(f"\\n=== æœºå™¨äººä¿¡æ¯ ===")
robots = [hero, infantry, engineer, hero2, infantry2]

for robot in robots:
    print(robot)  # è°ƒç”¨ __str__ æ–¹æ³•

print(f"\\næ€»æœºå™¨äººæ•°é‡: {Robot.get_robot_count()}")

# æµ‹è¯•æœºå™¨äººåŠŸèƒ½
print(f"\\n=== æœºå™¨äººæ“ä½œæµ‹è¯• ===")

# ç§»åŠ¨æµ‹è¯•
hero.move(100, 200)
infantry.move(150, 250)

# æˆ˜æ–—æµ‹è¯•
print(f"\\næˆ˜æ–—æ¨¡æ‹Ÿ:")
hero.take_damage(50)
infantry.take_damage(150)
engineer.take_damage(300)  # è‡´å‘½ä¼¤å®³

# ä¿®å¤æµ‹è¯•
print(f"\\nä¿®å¤æµ‹è¯•:")
hero.repair(30)
infantry.repair(100)

# çŠ¶æ€æŸ¥è¯¢
print(f"\\n=== æœºå™¨äººçŠ¶æ€æ€»è§ˆ ===")
for robot in robots:
    if robot.hp > 0:  # åªæ˜¾ç¤ºå­˜æ´»çš„æœºå™¨äºº
        status = robot.get_status()
        print(f"{status['name']}: {status['hp']}/{status['max_hp']} HP "
              f"({status['hp_percent']:.1f}%) at {status['position']}")

# é™æ€æ–¹æ³•æµ‹è¯•
print(f"\\n=== å·¥å…·å‡½æ•°æµ‹è¯• ===")
distance = Robot.calculate_distance((0, 0), (100, 200))
print(f"ä¸¤ç‚¹é—´è·ç¦»: {distance:.2f}")

valid_pos = Robot.is_valid_position(500, 600)
invalid_pos = Robot.is_valid_position(1500, 2000)
print(f"ä½ç½®(500, 600)æœ‰æ•ˆ: {valid_pos}")
print(f"ä½ç½®(1500, 2000)æœ‰æ•ˆ: {invalid_pos}")

print("\\nâœ… åŸºç¡€ç±»å®šä¹‰å’Œä½¿ç”¨å­¦ä¹ å®Œæˆï¼")
```

### 2.1 å°è£… (Encapsulation)

å°è£…æ˜¯å°†æ•°æ®å’Œæ“ä½œæ•°æ®çš„æ–¹æ³•ç»‘å®šåœ¨ä¸€èµ·ï¼Œå¹¶éšè—å†…éƒ¨å®ç°ç»†èŠ‚çš„ç‰¹æ€§ã€‚

```python
# å°è£…å’Œç»§æ‰¿ç¤ºä¾‹

print("=== é¢å‘å¯¹è±¡ä¸‰å¤§ç‰¹æ€§æ¼”ç¤º ===")

# åŸºç±»ï¼šæœºå™¨äºº
class Robot:
    """åŸºç¡€æœºå™¨äººç±»"""
    
    def __init__(self, name, robot_type, hp=200):
        self.name = name
        self.robot_type = robot_type
        self._hp = hp  # å—ä¿æŠ¤å±æ€§
        self._max_hp = hp
        self.position = (0, 0)
    
    @property
    def hp(self):
        return self._hp
    
    @hp.setter 
    def hp(self, value):
        self._hp = max(0, min(value, self._max_hp))
    
    def move(self, x, y):
        self.position = (x, y)
        print(f"{self.name} ç§»åŠ¨åˆ° ({x}, {y})")
    
    def attack(self) -> int:
        print(f"{self.name} æ‰§è¡ŒåŸºç¡€æ”»å‡»")
        return 30
    
    def take_damage(self, damage):
        self.hp -= damage
        print(f"{self.name} å—åˆ° {damage} ä¼¤å®³ï¼Œå‰©ä½™HP: {self.hp}")

# æ´¾ç”Ÿç±»ï¼šè‹±é›„æœºå™¨äºº
class HeroRobot(Robot):
    """è‹±é›„æœºå™¨äººç±» - ç»§æ‰¿ç¤ºä¾‹"""
    
    def __init__(self, name):
        super().__init__(name, "Hero", 400)  # è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        self.special_weapon = "42mmå¤§ç‚®"
        self.armor_level = 3
    
    def attack(self) -> int:
        """é‡å†™æ”»å‡»æ–¹æ³•"""
        print(f"{self.name} ä½¿ç”¨ {self.special_weapon} æ”»å‡»!")
        return 80
    
    def special_attack(self):
        """è‹±é›„ç‰¹æœ‰æŠ€èƒ½"""
        print(f"{self.name} å‘åŠ¨ç»ˆææŠ€èƒ½!")
        return 120

# æ´¾ç”Ÿç±»ï¼šæ­¥å…µæœºå™¨äºº  
class InfantryRobot(Robot):
    """æ­¥å…µæœºå™¨äººç±»"""
    
    def __init__(self, name):
        super().__init__(name, "Infantry", 200)
        self.weapon = "17mmæœºæª"
        self.mobility = "é«˜"
    
    def attack(self) -> int:
        """é‡å†™æ”»å‡»æ–¹æ³•"""
        print(f"{self.name} ä½¿ç”¨ {self.weapon} å¿«é€Ÿå°„å‡»!")
        return 50
    
    def quick_move(self, x, y):
        """æ­¥å…µç‰¹æœ‰æŠ€èƒ½"""
        self.move(x, y)
        print(f"{self.name} å¿«é€ŸæœºåŠ¨å®Œæˆ!")

# å·¥ç¨‹æœºå™¨äºº
class EngineerRobot(Robot):
    """å·¥ç¨‹æœºå™¨äººç±»"""
    
    def __init__(self, name):
        super().__init__(name, "Engineer", 250)
        self.tools = ["ä¿®å¤å·¥å…·", "å»ºé€ å·¥å…·"]
    
    def attack(self) -> int:
        """å·¥ç¨‹æœºå™¨äººçš„é˜²å¾¡æ€§æ”»å‡»"""
        print(f"{self.name} å‘å°„å¹²æ‰°å¼¹!")
        return 20
    
    def repair(self, target_robot):
        """å·¥ç¨‹æœºå™¨äººç‰¹æœ‰æŠ€èƒ½"""
        heal_amount = 50
        target_robot.hp += heal_amount
        print(f"{self.name} ä¿®å¤äº† {target_robot.name}ï¼Œæ¢å¤ {heal_amount} HP")

# æµ‹è¯•ç»§æ‰¿å’Œå¤šæ€
print("\\n=== ç»§æ‰¿å’Œå¤šæ€æµ‹è¯• ===")

# åˆ›å»ºä¸åŒç±»å‹çš„æœºå™¨äºº
robots = [
    HeroRobot("Hero-Alpha"),
    InfantryRobot("Infantry-Beta"), 
    InfantryRobot("Infantry-Gamma"),
    EngineerRobot("Engineer-Delta")
]

print("æœºå™¨äººé˜Ÿä¼:")
for robot in robots:
    print(f"  {robot.name} ({robot.robot_type}) - HP: {robot.hp}")

# å¤šæ€æ¼”ç¤ºï¼šä¸åŒæœºå™¨äººçš„æ”»å‡»æ–¹å¼
print(f"\\n=== å¤šæ€æ¼”ç¤ºï¼šæˆ˜æ–—æ¨¡æ‹Ÿ ===")
for robot in robots:
    damage = robot.attack()  # æ¯ä¸ªæœºå™¨äººéƒ½æœ‰è‡ªå·±çš„æ”»å‡»æ–¹å¼
    print(f"  é€ æˆä¼¤å®³: {damage}")

# ç‰¹æ®ŠæŠ€èƒ½æ¼”ç¤º
print(f"\\n=== ç‰¹æ®ŠæŠ€èƒ½æ¼”ç¤º ===")
hero = robots[0]
infantry = robots[1] 
engineer = robots[3]

# è‹±é›„ç‰¹æ®Šæ”»å‡»
hero.special_attack()

# æ­¥å…µå¿«é€Ÿç§»åŠ¨
infantry.quick_move(100, 150)

# å·¥ç¨‹æœºå™¨äººä¿®å¤
infantry.take_damage(80)  # å…ˆè®©æ­¥å…µå—ä¼¤
engineer.repair(infantry)  # å·¥ç¨‹æœºå™¨äººä¿®å¤

print("\\nâœ… ç»§æ‰¿å’Œå¤šæ€å­¦ä¹ å®Œæˆï¼")
```

## ç¬¬ä¸‰éƒ¨åˆ†ï¼šå¼‚å¸¸å¤„ç†ï¼ˆ20 åˆ†é’Ÿï¼‰

### 3.1 å¼‚å¸¸çš„åŸºæœ¬æ¦‚å¿µ

å¼‚å¸¸æ˜¯ç¨‹åºè¿è¡Œæ—¶å‘ç”Ÿçš„é”™è¯¯ï¼ŒPythonæä¾›äº†å®Œå–„çš„å¼‚å¸¸å¤„ç†æœºåˆ¶æ¥å¤„ç†è¿™äº›é”™è¯¯ã€‚

```python
# å¼‚å¸¸å¤„ç†è¯¦è§£

print("=== å¼‚å¸¸å¤„ç†æœºåˆ¶ ===")

# è‡ªå®šä¹‰å¼‚å¸¸ç±»
class RobotError(Exception):
    """æœºå™¨äººç›¸å…³å¼‚å¸¸çš„åŸºç±»"""
    pass

class RobotDestroyedException(RobotError):
    """æœºå™¨äººå·²æŸåå¼‚å¸¸"""
    pass

class InsufficientEnergyException(RobotError):
    """èƒ½é‡ä¸è¶³å¼‚å¸¸"""
    pass

class InvalidPositionException(RobotError):
    """æ— æ•ˆä½ç½®å¼‚å¸¸"""
    pass

class RobotControlSystem:
    """æœºå™¨äººæ§åˆ¶ç³»ç»Ÿ - å¼‚å¸¸å¤„ç†ç¤ºä¾‹"""
    
    def __init__(self):
        self.robots = {}
        self.field_size = (1000, 1000)
    
    def add_robot(self, robot_id, name, robot_type):
        """æ·»åŠ æœºå™¨äºº"""
        try:
            if robot_id in self.robots:
                raise ValueError(f"æœºå™¨äººID {robot_id} å·²å­˜åœ¨")
            
            hp_map = {"Hero": 400, "Infantry": 200, "Engineer": 250}
            if robot_type not in hp_map:
                raise ValueError(f"ä¸æ”¯æŒçš„æœºå™¨äººç±»å‹: {robot_type}")
            
            self.robots[robot_id] = {
                "name": name,
                "type": robot_type,
                "hp": hp_map[robot_type],
                "max_hp": hp_map[robot_type],
                "energy": 100,
                "position": (0, 0),
                "status": "æ­£å¸¸"
            }
            
            print(f"âœ… æˆåŠŸæ·»åŠ æœºå™¨äºº: {name} (ID: {robot_id})")
            
        except ValueError as e:
            print(f"âŒ æ·»åŠ æœºå™¨äººå¤±è´¥: {e}")
            raise
    
    def move_robot(self, robot_id, x, y):
        """ç§»åŠ¨æœºå™¨äºº"""
        try:
            # æ£€æŸ¥æœºå™¨äººæ˜¯å¦å­˜åœ¨
            if robot_id not in self.robots:
                raise KeyError(f"æœºå™¨äººID {robot_id} ä¸å­˜åœ¨")
            
            robot = self.robots[robot_id]
            
            # æ£€æŸ¥æœºå™¨äººçŠ¶æ€
            if robot["status"] == "æŸå":
                raise RobotDestroyedException(f"æœºå™¨äºº {robot['name']} å·²æŸåï¼Œæ— æ³•ç§»åŠ¨")
            
            # æ£€æŸ¥ä½ç½®æœ‰æ•ˆæ€§
            if not (0 <= x <= self.field_size[0] and 0 <= y <= self.field_size[1]):
                raise InvalidPositionException(f"ä½ç½® ({x}, {y}) è¶…å‡ºåœºåœ°èŒƒå›´")
            
            # æ£€æŸ¥èƒ½é‡
            energy_cost = 10
            if robot["energy"] < energy_cost:
                raise InsufficientEnergyException(f"èƒ½é‡ä¸è¶³ï¼Œéœ€è¦ {energy_cost}ï¼Œå½“å‰ {robot['energy']}")
            
            # æ‰§è¡Œç§»åŠ¨
            old_pos = robot["position"]
            robot["position"] = (x, y)
            robot["energy"] -= energy_cost
            
            print(f"âœ… {robot['name']} ä» {old_pos} ç§»åŠ¨åˆ° ({x}, {y})ï¼Œæ¶ˆè€—èƒ½é‡ {energy_cost}")
            
        except (KeyError, RobotDestroyedException, InvalidPositionException, InsufficientEnergyException) as e:
            print(f"âŒ ç§»åŠ¨å¤±è´¥: {e}")
            return False
        except Exception as e:
            print(f"âŒ æ„å¤–é”™è¯¯: {e}")
            return False
        
        return True
    
    def attack_robot(self, attacker_id, target_id):
        """æœºå™¨äººæ”»å‡»"""
        try:
            # æ£€æŸ¥æ”»å‡»è€…å’Œç›®æ ‡
            if attacker_id not in self.robots:
                raise KeyError(f"æ”»å‡»è€… {attacker_id} ä¸å­˜åœ¨")
            if target_id not in self.robots:
                raise KeyError(f"ç›®æ ‡ {target_id} ä¸å­˜åœ¨")
            
            attacker = self.robots[attacker_id]
            target = self.robots[target_id]
            
            # æ£€æŸ¥æ”»å‡»è€…çŠ¶æ€
            if attacker["status"] == "æŸå":
                raise RobotDestroyedException(f"æ”»å‡»è€… {attacker['name']} å·²æŸå")
            
            # æ£€æŸ¥ç›®æ ‡çŠ¶æ€
            if target["status"] == "æŸå":
                raise RobotDestroyedException(f"ç›®æ ‡ {target['name']} å·²æŸå")
            
            # æ£€æŸ¥èƒ½é‡
            if attacker["energy"] < 20:
                raise InsufficientEnergyException("æ”»å‡»éœ€è¦20ç‚¹èƒ½é‡")
            
            # æ‰§è¡Œæ”»å‡»
            damage = 50
            attacker["energy"] -= 20
            target["hp"] -= damage
            
            print(f"ğŸ’¥ {attacker['name']} æ”»å‡» {target['name']}ï¼Œé€ æˆ {damage} ä¼¤å®³")
            
            # æ£€æŸ¥ç›®æ ‡æ˜¯å¦è¢«æ‘§æ¯
            if target["hp"] <= 0:
                target["hp"] = 0
                target["status"] = "æŸå"
                print(f"ğŸ’€ {target['name']} è¢«æ‘§æ¯!")
            
        except Exception as e:
            print(f"âŒ æ”»å‡»å¤±è´¥: {e}")
            return False
        
        return True
    
    def get_robot_status(self, robot_id):
        """è·å–æœºå™¨äººçŠ¶æ€"""
        try:
            if robot_id not in self.robots:
                raise KeyError(f"æœºå™¨äººID {robot_id} ä¸å­˜åœ¨")
            
            return self.robots[robot_id].copy()
            
        except KeyError as e:
            print(f"âŒ æŸ¥è¯¢å¤±è´¥: {e}")
            return None

# å¼‚å¸¸å¤„ç†æµ‹è¯•
print("\\n=== å¼‚å¸¸å¤„ç†æµ‹è¯• ===")

system = RobotControlSystem()

# æµ‹è¯•æ­£å¸¸æ“ä½œ
try:
    system.add_robot("R001", "Hero-1", "Hero")
    system.add_robot("R002", "Infantry-1", "Infantry")
    print("\\næ­£å¸¸æ“ä½œå®Œæˆ")
except Exception as e:
    print(f"æ“ä½œå¤±è´¥: {e}")

# æµ‹è¯•å¼‚å¸¸æƒ…å†µ
print(f"\\n=== å¼‚å¸¸æƒ…å†µæµ‹è¯• ===")

# 1. é‡å¤IDå¼‚å¸¸
try:
    system.add_robot("R001", "Hero-2", "Hero")
except ValueError:
    print("âœ… æˆåŠŸæ•è·é‡å¤IDå¼‚å¸¸")

# 2. æ— æ•ˆæœºå™¨äººç±»å‹
try:
    system.add_robot("R003", "Unknown", "Tank")
except ValueError:
    print("âœ… æˆåŠŸæ•è·æ— æ•ˆç±»å‹å¼‚å¸¸")

# 3. ç§»åŠ¨åˆ°æ— æ•ˆä½ç½®
try:
    system.move_robot("R001", 2000, 3000)
except:
    print("âœ… æˆåŠŸæ•è·æ— æ•ˆä½ç½®å¼‚å¸¸")

# 4. èƒ½é‡ä¸è¶³ï¼ˆé€šè¿‡å¤šæ¬¡ç§»åŠ¨æ¶ˆè€—èƒ½é‡ï¼‰
print(f"\\n=== èƒ½é‡æ¶ˆè€—æµ‹è¯• ===")
for i in range(12):  # å°è¯•ç§»åŠ¨12æ¬¡ï¼Œæ¯æ¬¡æ¶ˆè€—10èƒ½é‡
    success = system.move_robot("R001", i*10, i*10)
    if not success:
        print(f"ç¬¬{i+1}æ¬¡ç§»åŠ¨å¤±è´¥")
        break

# 5. æ”»å‡»å·²æŸåçš„æœºå™¨äºº
print(f"\\n=== æˆ˜æ–—æµ‹è¯• ===")
# å…ˆæ”»å‡»ç›´åˆ°ç›®æ ‡è¢«æ‘§æ¯
for i in range(10):  # å¤šæ¬¡æ”»å‡»
    success = system.attack_robot("R001", "R002")
    if not success:
        break

# å°è¯•æ”»å‡»å·²æŸåçš„æœºå™¨äºº
system.attack_robot("R001", "R002")

print("\\nâœ… å¼‚å¸¸å¤„ç†å­¦ä¹ å®Œæˆï¼")
```

## ç¬¬å››éƒ¨åˆ†ï¼šç»¼åˆå®è·µé¡¹ç›® - æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨

è®©æˆ‘ä»¬åˆ›å»ºä¸€ä¸ªå®Œæ•´çš„æœºå™¨äººæ§åˆ¶ç³»ç»Ÿï¼Œç»¼åˆè¿ç”¨æ‰€æœ‰å­¦åˆ°çš„çŸ¥è¯†ã€‚

```python
# ç»¼åˆå®è·µé¡¹ç›®ï¼šæœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨

import json
import random
from datetime import datetime
from enum import Enum

print("ğŸ¤– RoboMaster æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨")
print("=" * 60)

class RobotState(Enum):
    """æœºå™¨äººçŠ¶æ€æšä¸¾"""
    STANDBY = "å¾…å‘½"
    ACTIVE = "æ¿€æ´»"
    MOVING = "ç§»åŠ¨ä¸­" 
    FIGHTING = "æˆ˜æ–—ä¸­"
    DESTROYED = "å·²æŸå"

class BattleRobot:
    """æˆ˜æ–—æœºå™¨äººåŸºç±»"""
    
    def __init__(self, robot_id, name, robot_type):
        self.robot_id = robot_id
        self.name = name
        self.robot_type = robot_type
        self.position = (0, 0)
        self.state = RobotState.STANDBY
        self.created_time = datetime.now()
        
        # æ ¹æ®ç±»å‹è®¾ç½®å±æ€§
        self._setup_attributes()
    
    def _setup_attributes(self):
        """æ ¹æ®æœºå™¨äººç±»å‹è®¾ç½®å±æ€§"""
        type_config = {
            "Hero": {"hp": 400, "energy": 120, "attack": 80, "speed": 2.5},
            "Infantry": {"hp": 200, "energy": 100, "attack": 50, "speed": 3.5},
            "Engineer": {"hp": 250, "energy": 150, "attack": 30, "speed": 3.0},
            "Drone": {"hp": 100, "energy": 80, "attack": 35, "speed": 5.0}
        }
        
        config = type_config.get(self.robot_type, type_config["Infantry"])
        self.max_hp = config["hp"]
        self.hp = config["hp"]
        self.max_energy = config["energy"]
        self.energy = config["energy"]
        self.attack_power = config["attack"]
        self.speed = config["speed"]
    
    def move_to(self, x, y):
        """ç§»åŠ¨åˆ°æŒ‡å®šä½ç½®"""
        if self.state == RobotState.DESTROYED:
            raise Exception(f"{self.name} å·²æŸåï¼Œæ— æ³•ç§»åŠ¨")
        
        old_pos = self.position
        distance = ((x - old_pos[0])**2 + (y - old_pos[1])**2)**0.5
        energy_cost = int(distance * 2)
        
        if self.energy < energy_cost:
            raise Exception(f"èƒ½é‡ä¸è¶³ï¼šéœ€è¦{energy_cost}ï¼Œå½“å‰{self.energy}")
        
        self.position = (x, y)
        self.energy -= energy_cost
        self.state = RobotState.MOVING
        
        return f"{self.name} ä»{old_pos}ç§»åŠ¨åˆ°({x}, {y})ï¼Œæ¶ˆè€—{energy_cost}èƒ½é‡"
    
    def attack(self, target):
        """æ”»å‡»ç›®æ ‡"""
        if self.state == RobotState.DESTROYED:
            raise Exception(f"{self.name} å·²æŸåï¼Œæ— æ³•æ”»å‡»")
        
        if target.state == RobotState.DESTROYED:
            raise Exception(f"ç›®æ ‡ {target.name} å·²æŸå")
        
        energy_cost = 20
        if self.energy < energy_cost:
            raise Exception("èƒ½é‡ä¸è¶³ï¼Œæ— æ³•æ”»å‡»")
        
        # è®¡ç®—ä¼¤å®³ï¼ˆåŠ å…¥éšæœºå› ç´ ï¼‰
        base_damage = self.attack_power
        damage = random.randint(int(base_damage * 0.8), int(base_damage * 1.2))
        
        self.energy -= energy_cost
        target.take_damage(damage)
        self.state = RobotState.FIGHTING
        
        return f"{self.name} æ”»å‡» {target.name}ï¼Œé€ æˆ {damage} ä¼¤å®³"
    
    def take_damage(self, damage):
        """å—åˆ°ä¼¤å®³"""
        self.hp -= damage
        if self.hp <= 0:
            self.hp = 0
            self.state = RobotState.DESTROYED
    
    def repair(self, amount):
        """ä¿®å¤"""
        if self.state != RobotState.DESTROYED:
            old_hp = self.hp
            self.hp = min(self.max_hp, self.hp + amount)
            return f"{self.name} ä¿®å¤ {self.hp - old_hp} HP"
        return f"{self.name} å·²æŸåï¼Œæ— æ³•ä¿®å¤"
    
    def recharge(self, amount):
        """å……èƒ½"""
        old_energy = self.energy
        self.energy = min(self.max_energy, self.energy + amount)
        return f"{self.name} å……èƒ½ {self.energy - old_energy}"
    
    def get_status(self):
        """è·å–çŠ¶æ€ä¿¡æ¯"""
        return {
            "id": self.robot_id,
            "name": self.name,
            "type": self.robot_type,
            "hp": f"{self.hp}/{self.max_hp}",
            "energy": f"{self.energy}/{self.max_energy}",
            "position": self.position,
            "state": self.state.value,
            "alive": self.state != RobotState.DESTROYED
        }

class BattleSimulator:
    """æˆ˜æ–—æ¨¡æ‹Ÿå™¨"""
    
    def __init__(self):
        self.robots = {}
        self.battle_log = []
        self.round_count = 0
    
    def add_robot(self, robot_id, name, robot_type):
        """æ·»åŠ æœºå™¨äºº"""
        if robot_id in self.robots:
            raise ValueError(f"æœºå™¨äººID {robot_id} å·²å­˜åœ¨")
        
        robot = BattleRobot(robot_id, name, robot_type)
        self.robots[robot_id] = robot
        
        log_entry = f"[å›åˆ{self.round_count}] æ·»åŠ æœºå™¨äºº: {name} ({robot_type})"
        self.battle_log.append(log_entry)
        return log_entry
    
    def execute_move(self, robot_id, x, y):
        """æ‰§è¡Œç§»åŠ¨æŒ‡ä»¤"""
        try:
            robot = self.robots[robot_id]
            result = robot.move_to(x, y)
            log_entry = f"[å›åˆ{self.round_count}] {result}"
            self.battle_log.append(log_entry)
            return log_entry
        except Exception as e:
            error_msg = f"[å›åˆ{self.round_count}] ç§»åŠ¨å¤±è´¥: {e}"
            self.battle_log.append(error_msg)
            return error_msg
    
    def execute_attack(self, attacker_id, target_id):
        """æ‰§è¡Œæ”»å‡»æŒ‡ä»¤"""
        try:
            attacker = self.robots[attacker_id]
            target = self.robots[target_id]
            
            result = attacker.attack(target)
            log_entry = f"[å›åˆ{self.round_count}] {result}"
            self.battle_log.append(log_entry)
            
            if target.state == RobotState.DESTROYED:
                destroy_msg = f"[å›åˆ{self.round_count}] ğŸ’€ {target.name} è¢«æ‘§æ¯!"
                self.battle_log.append(destroy_msg)
                log_entry += "\\n" + destroy_msg
            
            return log_entry
        except Exception as e:
            error_msg = f"[å›åˆ{self.round_count}] æ”»å‡»å¤±è´¥: {e}"
            self.battle_log.append(error_msg)
            return error_msg
    
    def get_alive_robots(self):
        """è·å–å­˜æ´»çš„æœºå™¨äºº"""
        return [robot for robot in self.robots.values() 
                if robot.state != RobotState.DESTROYED]
    
    def get_battle_summary(self):
        """è·å–æˆ˜æ–—æ€»ç»“"""
        alive_robots = self.get_alive_robots()
        destroyed_robots = [robot for robot in self.robots.values() 
                          if robot.state == RobotState.DESTROYED]
        
        return {
            "total_robots": len(self.robots),
            "alive_count": len(alive_robots),
            "destroyed_count": len(destroyed_robots),
            "rounds": self.round_count,
            "alive_robots": [robot.get_status() for robot in alive_robots],
            "destroyed_robots": [robot.get_status() for robot in destroyed_robots]
        }
    
    def advance_round(self):
        """æ¨è¿›å›åˆ"""
        self.round_count += 1
        
        # ä¸ºæ‰€æœ‰å­˜æ´»æœºå™¨äººæ¢å¤å°‘é‡èƒ½é‡
        for robot in self.get_alive_robots():
            robot.recharge(10)
            robot.state = RobotState.ACTIVE
    
    def display_status(self):
        """æ˜¾ç¤ºå½“å‰çŠ¶æ€"""
        print(f"\\n=== ç¬¬ {self.round_count} å›åˆçŠ¶æ€ ===")
        
        alive_robots = self.get_alive_robots()
        if alive_robots:
            print("å­˜æ´»æœºå™¨äºº:")
            for robot in alive_robots:
                status = robot.get_status()
                print(f"  {status['name']}: HP {status['hp']}, èƒ½é‡ {status['energy']}, "
                      f"ä½ç½® {status['position']}, çŠ¶æ€ {status['state']}")
        
        destroyed_robots = [r for r in self.robots.values() if r.state == RobotState.DESTROYED]
        if destroyed_robots:
            print(f"\\nå·²æŸåæœºå™¨äºº: {[r.name for r in destroyed_robots]}")

# è¿è¡Œæˆ˜æ–—æ¨¡æ‹Ÿ
def run_battle_simulation():
    """è¿è¡Œå®Œæ•´çš„æˆ˜æ–—æ¨¡æ‹Ÿ"""
    
    simulator = BattleSimulator()
    
    print("\\nğŸ¯ åˆå§‹åŒ–æˆ˜æ–—æ¨¡æ‹Ÿå™¨")
    
    # æ·»åŠ æœºå™¨äºº
    print(simulator.add_robot("H1", "Hero-Alpha", "Hero"))
    print(simulator.add_robot("I1", "Infantry-Beta", "Infantry"))
    print(simulator.add_robot("I2", "Infantry-Gamma", "Infantry"))
    print(simulator.add_robot("E1", "Engineer-Delta", "Engineer"))
    
    simulator.display_status()
    
    # æ¨¡æ‹Ÿå¤šå›åˆæˆ˜æ–—
    battle_actions = [
        # å›åˆ1ï¼šç§»åŠ¨å’Œå®šä½
        ("move", "H1", (100, 100)),
        ("move", "I1", (200, 150)),
        ("move", "I2", (150, 200)),
        ("move", "E1", (50, 50)),
        
        # å›åˆ2ï¼šå¼€å§‹æˆ˜æ–—
        ("attack", "H1", "I1"),
        ("attack", "I2", "H1"),
        ("move", "E1", (120, 120)),
        
        # å›åˆ3ï¼šç»§ç»­æˆ˜æ–—
        ("attack", "H1", "I2"),
        ("attack", "I1", "H1"),
        ("attack", "I2", "I1"),
    ]
    
    action_count = 0
    for action_type, robot_id, target in battle_actions:
        action_count += 1
        
        if action_count % 4 == 1:  # æ¯4ä¸ªåŠ¨ä½œæ¨è¿›ä¸€å›åˆ
            simulator.advance_round()
        
        if action_type == "move":
            result = simulator.execute_move(robot_id, target[0], target[1])
        elif action_type == "attack":
            result = simulator.execute_attack(robot_id, target)
        else:
            result = "æœªçŸ¥åŠ¨ä½œ"
        
        print(f"\\nåŠ¨ä½œ {action_count}: {result}")
        
        # æ£€æŸ¥æ˜¯å¦æœ‰æœºå™¨äººè¢«æ‘§æ¯
        alive_count = len(simulator.get_alive_robots())
        if alive_count <= 1:
            print(f"\\nğŸ æˆ˜æ–—ç»“æŸï¼åªå‰© {alive_count} ä¸ªæœºå™¨äººå­˜æ´»")
            break
    
    # æ˜¾ç¤ºæœ€ç»ˆçŠ¶æ€
    simulator.display_status()
    
    # æ˜¾ç¤ºæˆ˜æ–—æ€»ç»“
    summary = simulator.get_battle_summary()
    print(f"\\nğŸ“Š æˆ˜æ–—æ€»ç»“:")
    print(f"æ€»å›åˆæ•°: {summary['rounds']}")
    print(f"å‚æˆ˜æœºå™¨äºº: {summary['total_robots']}")
    print(f"å­˜æ´»: {summary['alive_count']}, æŸå: {summary['destroyed_count']}")
    
    if summary['alive_robots']:
        print(f"\\nğŸ† è·èƒœè€…:")
        for robot in summary['alive_robots']:
            print(f"  {robot['name']} - HP: {robot['hp']}")
    
    return simulator

# è¿è¡Œå®Œæ•´æ¨¡æ‹Ÿ
battle_simulator = run_battle_simulation()

print("\\n" + "=" * 60)
print("ğŸ‰ æœºå™¨äººæ§åˆ¶ç³»ç»Ÿæ¨¡æ‹Ÿå™¨æ¼”ç¤ºå®Œæˆ!")
print("=" * 60)

print(f"\\nğŸ“š æœ¬é¡¹ç›®ä½¿ç”¨çš„æŠ€æœ¯:")
print("âœ… é¢å‘å¯¹è±¡ç¼–ç¨‹ï¼ˆç±»ã€ç»§æ‰¿ã€å°è£…ï¼‰")
print("âœ… æšä¸¾ç±»å‹ï¼ˆRobotStateï¼‰") 
print("âœ… å¼‚å¸¸å¤„ç†ï¼ˆtry-exceptï¼‰")
print("âœ… åˆ—è¡¨å’Œå­—å…¸æ•°æ®ç»“æ„")
print("âœ… éšæœºæ•°ç”Ÿæˆ")
print("âœ… æ—¥æœŸæ—¶é—´å¤„ç†")
print("âœ… å­—ç¬¦ä¸²æ ¼å¼åŒ–")
print("âœ… æ¡ä»¶è¯­å¥å’Œå¾ªç¯")
print("âœ… å‡½æ•°è®¾è®¡å’Œæ¨¡å—åŒ–")

print(f"\\nğŸš€ æ­å–œå®Œæˆ Python åŸºç¡€è¯¾ç¨‹çš„å…¨éƒ¨å­¦ä¹ !")
print("ä½ å·²ç»æŒæ¡äº†Pythonç¼–ç¨‹çš„æ ¸å¿ƒæ¦‚å¿µï¼Œå¯ä»¥å¼€å§‹æ›´é«˜çº§çš„é¡¹ç›®å¼€å‘äº†ï¼")
```
